The Automobile Sales Dataset contains detailed sales data for an automobile company, including information on order details, product lines, and customer information. It features key columns such as quantity ordered, sales amount, and deal size, which can be used for sales trend analysis and customer behavior insights.


## **Source:**

## **Objective:**

The objective of this EDA is to understand which **product lines and deal sizes contribute the most to overall sales**.

```python
# Importing all required libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading the customer_bookings dataset

df = pd.read_csv("/content/drive/MyDrive/Auto Sales data.csv")
```

```python
df.head()
```

```python
# finding out the columns and rows in the dta set

df.shape
```

```python
# getting information about the values count, null values, data type of the field.

df.info()
```

```python
#I feel like summary statistics is very important in any EDA to understand about the dataset and its distribution.

df.describe()
```

```python
# Sales distribution histogram

df['SALES'].hist(bins=30)
plt.title('Sales Distribution')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

By observing above histogram, the highest sales amount frequency lies in between 2000$ - 4000$

```python
# Total sales by product line

sales_by_productline = df.groupby('PRODUCTLINE')['SALES'].sum()
sales_by_productline.plot(kind='bar')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

By looking at above bar chart, I can understand that only Classic Cars is giving highest sales among different product lines.

```python
# Sales distribution by deal size

sns.boxplot(x='DEALSIZE', y='SALES', data=df)
plt.title('Sales Distribution by Deal Size')
plt.show()
```

Even the deal size also, the medium deal size is more popular one.

```python
# Scatter plot of days since last order and sales

sns.scatterplot(x='DAYS_SINCE_LASTORDER', y='SALES', data=df, hue='DEALSIZE')
plt.title('Days Since Last Order vs Sales')
plt.show()
```

By comparing sales and deal sizes, the customers who are ordering medium deal size are more and the sales from that medium deals are more comparitively with other.

## **Conclusion:**

* More Sales is coming from Classic Cars in Product line.
* Medium Deal Size is providing more sales in Deal Size.





EDA helps us learn important things. It found that "Classic Cars" and medium deals make the most money. This tells us which products and deals are best for making more money.
