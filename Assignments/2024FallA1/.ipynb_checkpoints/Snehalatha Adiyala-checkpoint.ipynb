```python
import pandas as pd
house_data = pd.read_csv("House_data.csv")
house_data.head()
```

```python
house_data.info()
```

```python
house_data.isnull().sum()
```

```python
house_data.describe()
```

```python
#removing null values
house_data = house_data.dropna()
house_data.isnull().sum()
```

 Lets analyze house prices based on various features, such as the number of bedrooms, bathrooms, square footage, and location.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(house_data['price'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

The house prices are right-skewed, with most houses priced under $1 million. A few properties have very high prices, driving the tail of the distribution (as shown in the first histogram).

```python
# Distribution of number of bedrooms
plt.figure(figsize=(10,6))
sns.countplot(x=house_data['bedrooms'])
plt.title('Distribution of Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

The majority of houses have 3 or 4 bedrooms. Houses with more than 5 bedrooms are rare, and there is even a small number of houses with 1 or 2 bedrooms (as shown in the second bar plot).

```python
# Price vs. Square Footage of Living Area
plt.figure(figsize=(10,6))
sns.scatterplot(x=house_data['sqft_living'], y=house_data['price'])
plt.title('Price vs Square Footage of Living Area')
plt.xlabel('Square Footage of Living Area')
plt.ylabel('Price')
plt.show()
```

The scatter plot indicates a strong positive correlation between square footage and house price. Houses with larger living spaces tend to be more expensive, though there are a few outliers with very high prices even for smaller homes.

```python
# Price vs. Number of Bathrooms
plt.figure(figsize=(10,6))
sns.boxplot(x=house_data['bathrooms'], y=house_data['price'])
plt.title('Price vs Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price')
plt.show()
```

As the number of bathrooms increases, the house price tends to rise. The box plot shows this trend, where houses with more bathrooms generally have a higher price, but there are outliers (some expensive houses with fewer bathrooms).

```python
# Correlation matrix
plt.figure(figsize=(12,8))
corr_matrix = house_data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

sqft_living shows a strong correlation with price (0.7), indicating that houses with larger living spaces tend to be more expensive.

Bathrooms and grade also show a notable positive correlation with price, reinforcing that higher-quality homes and those with more bathrooms tend to be priced higher.

Interestingly, bedrooms have a weaker correlation with price than expected, suggesting that the size of the living area and the quality of the home (grade) may be more important factors in determining price.

In this EDA, I observed key factors that influence house prices in King County. Variables such as square footage of living area, number of bathrooms, and home grade were the most correlated with house prices. This analysis will guide us in making further decisions or applying machine learning models for predicting house prices.
