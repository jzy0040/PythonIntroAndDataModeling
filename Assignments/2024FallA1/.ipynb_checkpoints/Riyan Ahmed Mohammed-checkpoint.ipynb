```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/mohammedriyanahmed/Downloads/Dry_Bean_Dataset.csv')
```

```python
data.head()
```

```python
data.info()
```

## Objective
The objective of this EDA is to investigate the relationships between the physical characteristics of dry beans, such as area, perimeter, and compactness, to see how these features correlate with one another.

```python
data[['Area', 'Perimeter', 'Compactness']].describe()
```

## Descriptive Statistics for Area, Perimeter, and Compactness:

- Area: The mean area is around 53,048, but the high standard deviation indicates that the bean sizes vary a lot, with some beans being much larger.
- Perimeter: Similar to Area, the perimeter also shows variability, though to a lesser extent.
- Compactness: Compactness has much less variability compared to Area and Perimeter, meaning the beans are fairly consistent in this feature.
- The maximum values for Area and Perimeter are far higher than the 75th percentile, suggesting the presence of significant outliers.

These statistics confirm the observations made earlier about the skewness and variability in the dataset, especially for Area and Perimeter.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Area', y='Perimeter', data=data)
plt.title('Scatter Plot between Area and Perimeter')
plt.show()
```

### Scatter Plot Analysis

The scatter plot below illustrates the relationship between the **Area** and **Perimeter** of the dry beans. 

- From the plot, we can observe a strong positive correlation between these two variables. As the Area of the beans increases, the Perimeter also tends to increase, suggesting that larger beans typically have a longer perimeter.
- The scatter plot provides a clear visualization of how closely related these two variables are, further reinforcing the trend observed earlier in the correlation matrix.

```python
correlation_matrix = data[['Area', 'Perimeter', 'Compactness']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.show()
```

## Observations from Correlation Matrix

The correlation matrix provides insights into how the features relate to one another:

- Area and Perimeter have a very strong positive correlation (0.97). This indicates that as the size (area) of the bean increases, the perimeter also increases, which is expected.
- Compactness has a negative correlation with both Area (-0.27) and Perimeter (-0.41), meaning that larger beans tend to be less compact. However, the correlation is relatively weak, suggesting that the compactness of a bean is not strongly tied to its size or perimeter.
- The correlation values align with the visual patterns we observed in the pair plot, where Area and Perimeter exhibited a clear linear relationship, while Compactness showed more variability and independence from size.

This correlation analysis helps confirm the key relationships between the features and provides valuable insight into how these attributes interact.

```python
data[['Area', 'Perimeter', 'Compactness']].hist(bins=20, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

## Observations from Histogram Distribution:

The histograms show the distribution of the three key features—Area, Perimeter, and Compactness:
- **Area**: The distribution is right-skewed, indicating that most beans have smaller areas, but there are a few beans with much larger areas.
- **Perimeter**: Similar to the area, the perimeter is also right-skewed, which is expected as larger beans have larger perimeters.
- **Compactness**: This feature appears to be more normally distributed, with most beans having a compactness score around 0.85. The distribution is relatively symmetrical, showing consistency in compactness across the beans.

These distributions provide insights into the spread and variation within the dataset and highlight potential outliers in both the Area and Perimeter features.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Area'])
plt.title('Box Plot for Area')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Perimeter'])
plt.title('Box Plot for Perimeter')
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(y=data['Compactness'])
plt.title('Box Plot for Compactness')
plt.show()
```

### Observation from Box Plot Analysis for Area, Perimeter, and Compactness

To gain further insights into the distribution of data for **Area**, **Perimeter**, and **Compactness**, I have created individual box plots for each feature. This allows us to clearly see the spread, central tendency, and outliers of each feature without the issue of scaling problems from the combined box plot.

- **Box Plot for Area**: The area shows a large spread with a significant number of outliers. The data appears to be right-skewed, indicating that a few beans have much larger areas compared to the majority.
  
- **Box Plot for Perimeter**: Similar to the area, the perimeter also has a number of outliers and is right-skewed. Most beans have smaller perimeters, but a few beans stand out with much larger ones.

- **Box Plot for Compactness**: The compactness feature appears to be more symmetrically distributed, with fewer outliers compared to area and perimeter. The majority of the beans fall within a compactness range of 0.75 to 0.85, with a few outliers above 0.95.

These individual box plots provide clearer visualization of each feature's distribution and highlight the outliers more effectively, making it easier to interpret the variability within the dataset.

```python
from sklearn.linear_model import LinearRegression
X = data[['Area']].values
y = data['Perimeter'].values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Linear Regression: Area vs Perimeter')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.legend()
plt.show()
```

### Observation from Linear Regression: Area vs Perimeter

In this analysis, I performed a simple linear regression between the features "Area" and "Perimeter" to explore the relationship between them. The blue dots represent the actual data points, while the red line shows the fitted regression line.

- The regression line indicates the trend between the area and perimeter of the beans.
- From the graph, it is evident that there is a strong positive lnear relationship between Area and Perimeter, which was also confirmed by the correlation analysis.
- Most data points closely follow the regression line, confirming the strength of the linear relationship.

This regression model helps to quantify the relationship between these two features, and it could be useful for predictive purposes in further analysis.
