## Problem 1

A company wants to test whether the average time taken to assemble a product has significantly decreased after implementing a new training program. Prior to the training, the average assembly time was 35 minutes with a known population standard deviation of 5 minutes. After the training, a random sample of 40 employees showed a mean assembly time of 33 minutes. Can you help the company to decide whether the new training program is necessary?

```python
import pandas as pd
import numpy as np
from scipy import stats

# Data
mu0 = 35  # pre-training mean assembly time
sigma = 5  # population standard deviation
n = 40  # sample size
x_bar = 33  # post-training sample mean
alpha = 0.05  # significance level

# Step 1: Null and Alternative Hypotheses
# H0: μ = 35 (no significant decrease)
# H1: μ < 35 (significant decrease in assembly time)

# Step 2: Test statistic
z = (x_bar - mu0) / (sigma / (n ** 0.5))

# Step 3: Find critical value and decision rule
z_critical = norm.ppf(alpha)

# Step 4: Calculate the p-value
p_value = norm.cdf(z)

# Step 5: Conclusion
# Printing Results
print("Test Statistic (z):", z)
print("Critical Value (z_critical):", z_critical)
print("P-value:", p_value)

# Decision making by comparing Values
if z < z_critical:
    print("Reject H0: Significant decrease in assembly time.")
else:
    print("Fail to reject H0: No significant decrease in assembly time.")
```

## Problem 2
A university administrator wants to test whether graduate students at the institution study, on average, more than 25 hours per week. To explore this, a random sample of 15 graduate students was surveyed, and the sample mean study time was found to be 27 hours per week, with a sample standard deviation of 4.5 hours.

```python
from scipy.stats import t

# Data
mu0 = 25  # hypothesized mean study time
n = 15  # sample size
x_bar = 27  # sample mean
s = 4.5  # sample standard deviation
alpha = 0.05  # significance level

# Step 1: Null and Alternative Hypotheses
# H0: μ = 25 (no significant increase)
# H1: μ > 25 (significant increase)

# Step 2: Test statistic
t_stat = (x_bar - mu0) / (s / (n ** 0.5))

# Step 3: Find critical value and decision rule
t_critical = t.ppf(1 - alpha, n - 1)

# Step 4: Calculate the p-value
p_value = 1 - t.cdf(t_stat, n - 1)

# Step 5: Conclusion
# Printing Results
print("Test Statistic (t):", t_stat)
print("Critical Value (t_critical):", t_critical)
print("P-value:", p_value)

# Decesion making by comparing Values
if t_stat > t_critical:
    print("Reject H0: Graduate students study significantly more than 25 hours.")
else:
    print("Fail to reject H0: No significant increase in study time.")
```

**Output Expalanation :** Problem 2
Here’s a brief explanation of the output for Problem 2:

1. **Test Statistic (t-score)** **1.72** :
    Indicates the sample mean (27 hours) is 1.72 standard deviations above the hypothesized mean (25 hours).

2. **Critical Value (t_critical)** **1.76** :
   The threshold for rejecting the null hypothesis at a 0.05 significance level.

3. **P-value** **0.0536** :
   The probability of observing a t-score of 1.72 or more if the true mean is 25 hours.

Since the p-value (0.0536) is greater than the significance level (0.05), we **fail to reject the null hypothesis**. 
This suggests insufficient evidence to claim that graduate students study more than 25 hours per week on average.

```python
!pip install pulp   33#installing pulp
import pulp

#LP problem
lp_problem = pulp.LpProblem("MaximizeProduction", pulp.LpMaximize)

# Decision variables
xA = pulp.LpVariable('xA', lowBound=0, cat='Integer')  # units of Product A
xB = pulp.LpVariable('xB', lowBound=0, cat='Integer')  # units of Product B
xC = pulp.LpVariable('xC', lowBound=0, cat='Integer')  # units of Product C

# Objective function
lp_problem += xA + xB + xC, "TotalProductionOutput"

# Constraints
lp_problem += 2 * xA + xB + 3 * xC <= 100, "MachineXConstraint"
lp_problem += 4 * xA + 3 * xB + 2 * xC <= 85, "MachineYConstraint"

# Solveing the problem
lp_problem.solve()

# Display results
print("Optimal Solution:")
for variable in lp_problem.variables():
    print(f"{variable.name} = {variable.varValue}")

print("\nOptimal Objective Function Value:")
print(f"Z = {pulp.value(lp_problem.objective)}")
```
