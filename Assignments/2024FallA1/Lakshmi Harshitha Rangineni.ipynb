```python
laptop_prices = data.drop(['SecondaryStorage'], axis=1)
```

```python
laptop_prices.head()
```

```python
#realtionship analysis
```

```python
numeric_data = laptop_prices.select_dtypes(include=['number'])
```

```python
cleaned_data = numeric_data.dropna()
```

```python
corelation = cleaned_data.corr()
```

```python
#Displays heatmaps 
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```

```python
#displays pairplot
sns.pairplot(laptop_prices)
```

```python
sns.relplot(x='Company', y='Product', hue='TypeName', data=laptop_prices)
```

```python
#displays bar graph
sns.distplot(laptop_prices['Inches'])
```

```python
#displays boxplot 
sns.catplot(x='Inches' , kind= 'box', data= laptop_prices)
```
