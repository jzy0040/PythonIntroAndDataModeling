```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

# MY Dataset


```python
file_path = 'C:\Python Task\EDA\Chicaco Energy Benchmarking.xlsx'
data = pd.read_excel(file_path)
```

# Information about the dataset

```python
# Print basic information about the dataset
print("Dataset Information:")
print(data.info())

# Print the first few rows of the dataset
print("\nFirst Few Rows:")
print(data.head())

# Print statistical summary of the dataset
print("\nStatistical Summary:")
print(data.describe())
```

objective = "Objective: Analyze energy usage patterns across different types of properties in Chicago to identify opportunities for energy savings."
print("\n" + objective)


# Print the objective of the EDA

```python
# Histograms for each numeric column to visualize distributions
print("\nHistograms of Numeric Columns:")
data.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()

# Boxplots to identify outliers in energy usage data
print("\nBoxplots for Energy Use and Emissions:")
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[['Electricity Use (kBtu)', 'Natural Gas Use (kBtu)', 'Total GHG Emissions (Metric Tons CO2e)']])
plt.title('Boxplot of Energy Use and Emissions')
plt.show()

# Filter out non-numeric columns before computing the correlation matrix
numeric_data = data.select_dtypes(include=[np.number])  # This ensures only numeric columns are included

print("\nCorrelation Heatmap:")
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()


```
