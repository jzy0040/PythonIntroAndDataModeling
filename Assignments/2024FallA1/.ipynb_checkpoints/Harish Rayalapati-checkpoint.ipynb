```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
file_path = r"C:\Users\Harish\Downloads\heart.csv"
df = pd.read_csv(file_path)

# Overview of the dataset
print("Dataset Overview:")
print(df.info())  # Provides summary of the dataset including missing values and data types
print("\nFirst 5 rows of the dataset:")
print(df.head())  # Display first 5 rows

# Summary statistics
print("\nSummary statistics:")
print(df.describe())  # Provides summary statistics of numeric columns

# Objective: Understand which factors contribute the most to heart disease

# 1. Check for missing values
print("\nChecking for missing values:")
print(df.isnull().sum())  # Summarize missing values per column

# 2. Univariate Analysis - Distributions of Numeric Columns
# Histogram for age, cholesterol, blood pressure
df.hist(bins=20, figsize=(10, 10), color='blue')
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.show()

# Box plot for visualizing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title("Box Plot of All Numeric Columns", fontsize=16)
plt.xticks(rotation=90)
plt.show()

# 3. Bivariate Analysis - Relationship between Features and Heart Disease
# Correlation matrix
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap", fontsize=16)
plt.show()

# Scatter plot: Age vs Cholesterol color-coded by target (Heart Disease)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='chol', hue='target', data=df, palette='coolwarm')
plt.title("Age vs Cholesterol by Heart Disease Status", fontsize=16)
plt.show()

# Pair plot for all numeric variables colored by the target
sns.pairplot(df, hue='target', diag_kind='kde')
plt.suptitle("Pair Plot of Variables by Heart Disease Status", fontsize=16, y=1.02)
plt.show()

# 4. Multivariate Analysis: Feature Importance
# Let's identify important features using correlation with the target variable
corr_with_target = df.corr()['target'].sort_values(ascending=False)
print("\nCorrelation of Features with Heart Disease:")
print(corr_with_target)

# 5. Insights and Summary
# Key findings can be summarized in this section of your report.
# Example:
# - Age and cholesterol levels show a moderate positive correlation with heart disease.
# - Higher maximum heart rate achieved is negatively correlated with heart disease (i.e., lower risk).
```
