```python
import pandas as pd # imported libraries that are needed to generte code.
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/drive/MyDrive/breast-cancer.csv')#This function used to load dataset into a dataframe.
```

```python
print(data.head()) # Used to get an initial glimpse of the first few rows of the datset.
```

```python
print(data.tail()) # Display last few rows of the dataset.
```

```python
print(data.info()) # Display basic information about the dataset.
```

```python
print(data.describe()) # Display summary statistics of the dataset.
```

```python
#The purpose of this code is to visualize the distribution of the diagnosis variable in the dataset.
#The count plot will show how many samples are classified as malignant (M) and how many are classified as benign (B).
# This data can be used to analyze the features such as texture or size to make a diagnosis.

plt.figure(figsize=(8, 6))
sns.countplot(x='diagnosis', data=data)
plt.title('Distribution of Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
def univariate_analysis(data):
    # Display summary statistics
    print("Summary Statistics:")
    print(data.describe())

    # Plot histograms for numerical features
    data.hist(bins=20, figsize=(20, 15))
    plt.suptitle('Histograms of Numerical Features')
    plt.show()

    # Plot count plot for the target variable 'diagnosis'
    plt.figure(figsize=(8, 6))
    sns.countplot(x='diagnosis', data=data)
    plt.title('Distribution of Diagnosis')
    plt.xlabel('Diagnosis')
    plt.ylabel('Count')
    plt.show()

univariate_analysis(data)
```

```python
def bivariate_analysis(data):
    # Plot pairplot to visualize relationships between features
    sns.pairplot(data, hue='diagnosis', diag_kind='kde')
    plt.suptitle('Pairplot of Features', y=1.02)
    plt.show()

    # Plot correlation matrix
    plt.figure(figsize=(16, 12))
    correlation_matrix = data.corr()
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

bivariate_analysis(data)
```
