# Exploratory Data Analysis
## Student Performance Factors

### The objective of the EDA is to understand the factors affecting student performance and to identify trends that could help educators improve academic outcomes.

```python
#remove rows that contain NULL values

data.dropna(subset = ['Teacher_Quality','Parental_Education_Level','Distance_from_Home'], inplace=True)
```

```python
data.isnull().sum()
```

```python
# Create scatter plot to compare study hour vs exam scores
sns.scatterplot(x=data["Hours_Studied"],y=data["Exam_Score"])
plt.title("Hours vs Score Percentage")
plt.xlabel("Hours Studied")
plt.ylabel("Percentage Scored")
plt.show()
```

```python
data.loc[data['Exam_Score']>100, 'Exam_Score'] =100
```

```python
data["Gender"].value_counts()
```

```python
# Create a bar chart to explore gender
sns.countplot(data["Gender"])
```

```python
average_score_by_gender=data.groupby('Gender')['Exam_Score'].mean()
```

```python
# create bar chart to compare gender to average exam score

average_score_by_gender.plot(kind='bar')

plt.title('Average Exam Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Exam Score')
plt.show()
average_score_by_gender
```

```python
#create line chart to see if attendance has an impact on the average exam score

average_score_vs_attendance=data.groupby('Attendance')['Exam_Score'].mean()
average_score_vs_attendance.plot(kind='line')                                         
plt.title ('Attendance vs Exam Score')
plt.xlabel ('Attendance')
plt.ylabel ('Average Exam Score')
plt.show()
```

```python
high_attendance=data [(data['Attendance'] >=90)]
high_attendance
```

```python
#create scatter plotter chart to compare hours studied vs exam score

plt.scatter(data['Hours_Studied'],data['Exam_Score'])
plt.title ('Hours Studied vs Exam Score')
plt.xlabel ('Hours Studied')
plt.ylabel ('Average Exam Score')
plt.show()
```

```python
high_score_low_hours = data[(data['Hours_Studied'] <=1) & (data['Exam_Score'] >= 90)]
high_score_low_hours
```

```python
data.head()
```

```python
average_score_by_School_Type = data.groupby('School_Type')['Exam_Score'].mean()
```

```python
#create bar graph to compare average exam score vs school type 

average_score_by_School_Type.plot(kind='bar')
plt.title('Average Exam Score vs School Type')
plt.xlabel('School Type')
plt.ylabel('Average Exam Score')
plt.show()
average_score_by_School_Type
```

```python
average_score_by_Parental_Involvement = data.groupby('Parental_Involvement')['Exam_Score'].mean()
```

```python
#Create bar graph to determine if parental involvement has a impact of the student average exam score 

average_score_by_Parental_Involvement.plot(kind='bar')
plt.title('Average Exam Score by Parental Involvement')
plt.xlabel('Parental Involvement')
plt.ylabel('Average Exam Score')
plt.show()
average_score_by_Parental_Involvement
```

### Through EDA, I was able to identify key factors influencing student performance, specifically that attendance emerged as a significant predictor of average grades. By analyzing attendance data concerning grades, I found a clear positive correlation: increased attendance was associated with higher average grades. This insight shows the importance of consistent school attendance in enhancing academic performance. Future studies could explore the reason behind attendance patterns and develop strategies to improve attendance.
