# EDA Analysis

```python
#1
df.isnull().sum()
```

# From, the analysis we can clearly see that the dataset has no nulls.

```python
#I want to know that how many fraud transactions are there, and how many non fraud transactions are present.
df['Class'].value_counts()
```

## from the results it is evident that, there are almost 2,84,315 non fraud transactions and very little amount of fraud transactions i.e., 492 transactions

```python
#2
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the histogram with different colors for each class
sns.displot(data=df, x='Class', kind='hist', hue='Class', palette=['skyblue', 'salmon'], discrete=True)

# Set the y-axis to logarithmic scale
plt.yscale('log')

# Show the plot
plt.show()
```

### From, the above graph we can say that the dataset is highly imbalanced

# Correlation Matrix- which explains how 2 variables are related to each other

```python
#3 
df.corr()
```

```python
plt.figure(figsize=(20, 15))  
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

### From, the graph we can see that there are some variables which are having no relation to other vatiables and there are some variables which are very less correlated to each other

# Variance 

```python
variances = df.var()
print(variances)
```

# outlier detection

```python
sns.boxplot(x='Class',y='Amount',data=df)
plt.title('Boxplot of Transaction vs Amount')
plt.show()
```

### From the graph we can say that mostly fraud transactions are focused on transaction value which are less than 5000
