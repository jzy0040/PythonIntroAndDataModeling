```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
from scipy.stats import zscore
%matplotlib inline
```

### 	A company wants to test whether the average time taken to assemble a product has significantly decreased after implementing a new training program. Prior to the training, the average assembly time was 35 minutes with a known population standard deviation of 5 minutes. After the training, a random sample of 40 employees showed a mean assembly time of 33 minutes. Can you help the company to decide whether the new training program is necessary?

```python
population_mean = 35
population_std_dev = 5
sample_mean = 33
n = 40

z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(n))
p_value = stats.norm.cdf(z_score) 

print(f"Z-score: {z_score}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The training program has significantly decreased the assembly time.") 
else: 
    print("Fail to reject the null hypothesis. The training program has not significantly decreased the assembly time.")
```

**If the p-value is less than the significance level (0.05), reject the null hypothesis.**


1. **If the p-value is low (e.g., below 0.05), there is evidence that the training program significantly reduced the average assembly time.**
2. **If the p-value is not low, we fail to reject the null hypothesis, meaning there is insufficient evidence to conclude that the training program reduced assembly time.**

## 2.	A university administrator wants to test whether graduate students at the institution study, on average, more than 25 hours per week. To explore this, a random sample of 15 graduate students was surveyed, and the sample mean study time was found to be 27 hours per week, with a sample standard deviation of 4.5 hours.

```python
data = {
    'Hours_of_Study': [5, 8, 5, 7, 7, 8, 3, 8, 0, 8, 5, 0, 2, 8, 7, 7, 4, 4, 2, 6, 1, 7, 6, 2, 8, 0, 5, 8, 7, 4],
    'Score': [52.1221, 82.1221, 52.1221, 72.1221, 72.1221, 82.1221, 32.1221, 82.1221, 2.1221, 82.1221, 
              52.1221, 2.1221, 22.1221, 82.1221, 72.1221, 72.1221, 42.1221, 42.1221, 22.1221, 62.1221, 
              12.1221, 72.1221, 62.1221, 22.1221, 82.1221, 2.1221, 52.1221, 82.1221, 72.1221, 42.1221]
}
df = pd.DataFrame(data)

# Independent variable (X) and dependent variable (Y)
X = sm.add_constant(df['Hours_of_Study'])
Y = df['Score']

# Fit linear regression model
model = sm.OLS(Y, X).fit()

# Print model summary
print(model.summary())
```

```python
model.summary2()
```

1. **The Model R-Squared value is 1.000, that is model performance is good R-Squared value lies between 0 to 1, higher R-Squared indicates better fit**
2. **the p-value for t-test is 0.0000, which indicates there is statistically significant relationship (significane value is p = 0.05) between features Hours of study and score**
3. **Probability value of F-statistic is 1.7 that indicate overall model is statistical significant.**

```python
model.params
```

```python
estimate_prediction = 2.1221 + 10.0000 * 0.1
estimate_prediction
```

```python
model_res = model.resid
probplot = sm.ProbPlot(model_res)
plt.figure(figsize=(8, 6))
probplot.ppplot(line='45')
plt.title("Normal P-P Plot of Regression Standards Residuals")
plt.show()
```

```python
plt.scatter(df['Hours_of_Study'], df['Score'], color="blue", label="Data Points")
plt.plot(df['Hours_of_Study'], model.predict(X), color="red", label="Regression Line")
plt.xlabel("Hours of Study")
plt.ylabel("Exam Score")
plt.legend()
plt.show()
```

**Data points falls on regression line, Model prediction value are good fit**

```python
# Residual plot
residuals = model.resid
plt.scatter(model.predict(X), residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.show()

# Q-Q plot
sm.qqplot(residuals, line='45')
plt.title("Q-Q Plot of Residuals")
plt.show()

# Durbin-Watson test
durbin_watson_stat = sm.stats.stattools.durbin_watson(residuals)
print(f"Durbin-Watson Statistic: {durbin_watson_stat}")
```

**QQ plots are commonly used to compare a data set to a theoretical model.**
1. **This can provide an assessment of goodness of fit that is graphical, rather than reducing to a numerical summary statistic.**
2. **QQ plots are also used to compare two theoretical distributions to each other.**

**Durbin-Watson Statistic: a statistical test that detects autocorrelation in the residuals of a regression analysis
statistic has a value range of 04, with an acceptable range of 1.502.50:**
- **2 or nearly 2: Indicates no first-order autocorrelation**
- **Less than 1.50: Indicates positive autocorrelation**
- **More than 2.50: Indicates negative autocorrelation**

##### Test of Homoscedasticity

```python
def get_standardized_value(values):
    return (values - values.mean()) / values.std()
```

## Linear Optimization

```python
from scipy.optimize import linprog

# Coefficients of the objective function (negative for maximization)
c = [-1, -1, -1]  # Coefficients for maximizing production

# Coefficients for the inequality constraints
A = [
    [2, 1, 3],  # Machine X hours constraint
    [4, 3, 2]   # Machine Y hours constraint
]

# Right-hand side values for each constraint
b = [100, 85]  # Available hours for Machine X and Machine Y

# Solve the linear program
result = linprog(c, A_ub=A, b_ub=b, bounds=(0, None), method="highs")

# Output results
if result.success:
    print("Optimal Production Plan:")
    print(f"Product A: {result.x[0]} units")
    print(f"Product B: {result.x[1]} units")
    print(f"Product C: {result.x[2]} units")
    print(f"Maximum Total Production: {-result.fun}")
else:
    print("Optimization did not succeed.")
```

The optimal production plan based on the given constraints suggests the following:

1 **Product A**: Produce **0 units**.
2 **Product B**: Produce approximately **7.86 units**.
3 **Product C**: Produce approximately **30.71 units**.
4 **Maximum Total Production**: About **38.57 units**.

### Analysis

1. **Binding Constraint**:
   - To determine if Machine X or Machine Y is the bottleneck, we can substitute the optimal values back into the constraints:
     - **Machine X Usage**: (2(0) + 1(7.86) + 3(30.71) = 100) hours
     - **Machine Y Usage**: (4(0) + 3(7.86) + 2(30.71) = 85) hours
   - Both constraints are exactly met, meaning both machines are fully utilized. This indicates that **both Machine X and Machine Y are binding constraints** in this case.

2. **Upgrade Recommendation**:
   - Since both Machine X and Machine Y are fully utilized, an upgrade to either machine would increase production capacity. However:
     - **Machine X** is more critical for producing Product C (the highest production quantity in the optimal plan) as it requires 3 hours of Machine X time per unit, while Machine Y usage per unit is lower.
     - If budget allows only one machine upgrade, **Machine X** would likely be more impactful in increasing the total output, as it has the higher constraint coefficient for Product C.

Upgrading Machine X will enable the company to produce more of the high-demand products and potentially increase total production beyond 38.57 units.
