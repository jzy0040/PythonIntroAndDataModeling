```python
Analysis: I tried using Bivarate analysis to analyze the relationship between 'Schooling' and 'Life expectancy' in various countries. To make the datastudy 
a bit easier I selected 30 random samples from the dataset. The purpose of using random sample is to reduce the chances of any biasness that may occur.
Those samples were than plotted inorder to show the graphical representation of the values of those two variables. 
In the figure we can see that as the years of schooling increases the life expectancy also gradually increases. For some countries , even though the schooling
years are low the life expectancy can be seen a bit higher and vice versa. But those look like few case of exceptions . For majority of the countries we
can see positive relationship between both the variables, i.e. they move towards same direction.
```

```python
Use of EDA: Like most of the dataset, the dataset that I used for this assignment also contains 100s of samples. It is almost impossible to go through 
each and every sample individually to figure out what in the dataset are we looking to find. In situations like this, EDA comes in very handy. For my 
objective of finding out the influence of one variable on other, the visual representation of the values helped me easily read their pattern as oppose to 
manually going through the whole collection of data which would literally take forever. 
Here, I have used the most basic form of analysis but there are various other analytical techniques through which we can do indepth analysis. We can clean data
identify outliers, create hypothesis, findout the correlation and do numerous other things when we use other advance forms of analysis as well. 
It provides furthur scope of research and we can make informed decision on it's basis.
```
