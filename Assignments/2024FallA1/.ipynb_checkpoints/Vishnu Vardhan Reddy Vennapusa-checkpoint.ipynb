```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('C:/Users/vvvar/Downloads/hotel_booking.csv')
```

```python
# Display the info about the dataset
print(df.info())
```

```python
# Check for missing values
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0])
```

```python
df['children'] = df['children'].fillna(0)
```

```python
# Summary statistics of numeric columns
print(df.describe())
```

```python
# Plot the distribution of lead_time
plt.figure(figsize=(10,6))
sns.histplot(df['lead_time'], kde=True)
plt.title('Distribution of Lead Time')
plt.xlabel('Lead Time (Days)')
plt.show()
```

```python
# Plot cancellations by hotel type
plt.figure(figsize=(8,6))
sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Set2')
plt.title('Cancellations by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Plot number of bookings per month
plt.figure(figsize=(10,6))
sns.countplot(x='arrival_date_month', data=df, order=['January', 'February', 'March', 'April', 'May', 'June', 
                                                      'July', 'August', 'September', 'October', 'November', 'December'])
plt.title('Bookings by Month')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=[np.number])

# Correlation matrix for numeric columns
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Count plot for hotel types
plt.figure(figsize=(8,6))
sns.countplot(x='hotel', data=df, palette='Set2')
plt.title('Count of Bookings by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Box plot for lead_time vs is_canceled
plt.figure(figsize=(10,6))
sns.boxplot(x='is_canceled', y='lead_time', data=df, palette='Set1')
plt.title('Lead Time vs Cancellations')
plt.xlabel('Canceled (0 = No, 1 = Yes)')
plt.ylabel('Lead Time (Days)')
plt.show()
```

```python
# Count plot for customer types
plt.figure(figsize=(10,6))
sns.countplot(x='customer_type', data=df, palette='Set3')
plt.title('Booking Distribution by Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Box plot for ADR vs Hotel Type
plt.figure(figsize=(10,6))
sns.boxplot(x='hotel', y='adr', data=df, palette='muted')
plt.title('Average Daily Rate (ADR) by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```

```python
# Count plot for market segment
plt.figure(figsize=(12,6))
sns.countplot(x='market_segment', data=df, palette='Set2')
plt.title('Bookings by Market Segment')
plt.xticks(rotation=45)
plt.xlabel('Market Segment')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# Joint distribution of stays_in_week_nights and stays_in_weekend_nights
plt.figure(figsize=(10,6))
sns.jointplot(x='stays_in_week_nights', y='stays_in_weekend_nights', kind='scatter', data=df, height=8, color='g')
plt.title('Week Nights vs Weekend Nights Stay')
plt.show()
```

```python
# Cancellations by month
pivot_table = df.pivot_table(values='is_canceled', index='arrival_date_month', columns='hotel', aggfunc='sum')
pivot_table = pivot_table.reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'])

plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Cancellations by Month and Hotel Type')
plt.ylabel('Month')
plt.show()
```

```python
# Distribution plot for ADR
plt.figure(figsize=(10,6))
sns.histplot(df['adr'], kde=True, color='blue')
plt.title('Distribution of Average Daily Rate (ADR)')
plt.xlabel('Average Daily Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of Lead Time vs ADR
plt.figure(figsize=(10,6))
sns.scatterplot(x='lead_time', y='adr', hue='hotel', data=df)
plt.title('Lead Time vs ADR')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()
```
