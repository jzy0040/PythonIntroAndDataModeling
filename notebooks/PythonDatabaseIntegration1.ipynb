```python
import pandas as pd
import numpy as np

import mysql.connector

from getpass import getpass
```

```python
# Connect to MySQL server
conn = mysql.connector.connect(
    host = "localhost",      # means the database is on the same machine as your Python script. If the database is remote, replace it with the server's IP address (e.g., "192.168.1.100").
    user = "root",           # MySQL username; "root" is the default MySQL administrator user.
    password = getpass("Enter MySQL password: "), # User enter MySQL server password
)

cursor = conn.cursor() # an object that allows you to: Execute SQL commands and retrieve results from a MySQL database using Python.

# Create a database
cursor.execute("CREATE DATABASE IF NOT EXISTS MyFirstPythonDatabase") ### this line create the actual database

print("Database created successfully!")
# Close connection
conn.close()
```

```python
# Create a table

conn = mysql.connector.connect(
    host = "localhost",      
    user = "root",          
    password = getpass("Enter MySQL password: "), 
    database = "MyFirstPythonDatabase")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(100),
    email VARCHAR(100) UNIQUE
)
""")

print("Table created successfully!")
# Close connection
conn.close()
```

```python
# Insert data

conn = mysql.connector.connect(
    host = "localhost",      
    user = "root",          
    password = getpass("Enter MySQL password: "), 
    database = "MyFirstPythonDatabase")

cursor = conn.cursor()

cursor.execute("INSERT INTO users (name, email) VALUES (%s, %s)", ("Alice", "alice@example.com")) # %s is a place holder that allows you specify the actual value later 
cursor.execute("INSERT INTO users (name, email) VALUES (%s, %s)", ("Bob", "bob@example.com"))
cursor.execute("INSERT INTO users (name, email) VALUES (%s, %s)", ("Jack", "jack@example.com"))
cursor.execute("INSERT INTO users (name, email) VALUES (%s, %s)", ("Mike", "mike@example.com"))
cursor.execute("INSERT INTO users (name, email) VALUES (%s, %s)", ("Tom", "tom@example.com"))
conn.commit()   # Save changes

# Query data
cursor.execute("SELECT * FROM users")
for row in cursor.fetchall():
    print(row)

# Close connection
conn.close()
```
