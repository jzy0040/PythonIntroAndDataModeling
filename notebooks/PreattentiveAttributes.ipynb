# Preattentive Attributes

It's the visual properties that are processed by the brain very rapidly and without conscious effort. These attributes can help in guiding attention and highlighting important information in visualizations. Some common preattentive attributes include:

<span style='color:blue'>Color</span>: Differences in color can be perceived quickly, making it useful for encoding categories or highlighting specific data points.

<span style='color:blue'>Size</span>: Differences in size can also be perceived rapidly, helping to indicate magnitude or importance.

<span style='color:blue'>Shape</span>: Different shapes can be used to encode categories or distinguish between different types of data.

<span style='color:blue'>Orientation</span>: Changes in orientation, such as slope or angle, can be detected quickly and used to convey information.

<span style='color:blue'>Movement</span>: Dynamic changes or motion can attract attention and convey changes over time.

<span style='color:blue'>Spatial Position</span>: Differences in position along a common scale or axis can be perceived rapidly and used to compare values.

These examples showcase how preattentive attributes can be used to encode data in visualizations to make patterns and trends more easily perceptible.

```python
import numpy as np
import yfinance as yf
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# Sample data
np.random.seed(0)
x = np.arange(10)
y = np.random.randint(1, 10, size=10)
sizes = np.random.randint(10, 100, size=10)
colors = np.random.rand(10)

# Create a figure with multiple subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Color
axs[0, 0].scatter(x, y, s=100, c=colors, alpha=0.5)
axs[0, 0].set_title('Color')

# Size
axs[0, 1].scatter(x, y, s=sizes*10, c='blue', alpha=0.5)
axs[0, 1].set_title('Size')

# Shape
axs[1, 0].scatter(x, y, s=100, c='blue', alpha=0.5, marker='s')
axs[1, 0].set_title('Shape')

# Position
axs[1, 1].scatter(x, y, s=100, c='blue', alpha=0.5)
axs[1, 1].scatter(x+0.5, y, s=100, c='red', alpha=0.5)
axs[1, 1].set_title('Position')

plt.tight_layout()
plt.show()
```
