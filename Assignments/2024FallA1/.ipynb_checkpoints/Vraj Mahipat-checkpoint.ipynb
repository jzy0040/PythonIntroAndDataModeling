```python
# import the required modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns

'''
The dataset comprises 100,000 rows by 20 columns of data that highlight the key aspects of the employees,
productivity, and demographics of a corporate environment.
It contains information such as the employee's job, work habits, education, performance, and satisfaction.

'''
# load the csv 
df = pd.read_csv('Extended_Employee_Performance_and_Productivity_Data.csv') # source -> https://www.kaggle.com/datasets/mexwell/employee-performance-and-productivity-data

print(df.shape) #print the shape/dimensions of the dataset
df.head() # Get first five rows of the dataset
```

```python
# display statistical informartion about the dataset
df.info()
df.describe()
```

```python
# null check
df.isna().sum()
```

# Correlation Heatmap

```python
corr_matrix = df[df.select_dtypes(include=['int64', 'float64']).columns].corr() #compute correlation matrix
plt.figure(figsize=(12, 10)) #set the size of the canvas
sns.heatmap(corr_matrix, annot=True, linewidths=0.5) 
plt.title('Correlation Heatmap') # title
plt.show() # display the plot
```

 # Gender distribution

```python
gender_counts = df['Gender'].value_counts() # compute values for each gender

# set the colors of the partitions of the piechart
colors = ['orange', 'lightgreen' , 'gold']

explode = (0, 0.1, 0) # make the third partition pop out

# Gender count plot
plt.pie(gender_counts, explode=explode, labels=gender_counts.index,
        colors=colors, autopct='%1.1f%%', shadow=False, startangle=90)

plt.axis('equal')
plt.title('Pie Chart of Gender Distribution')
plt.show() #show the plot
```

Number of men and female are equal suggesting that the company takes gender equality seriously

# Distribution of Employee Satisfaction Scores

```python
plt.figure(figsize=(10, 6)) # set the figure size
sns.histplot(df['Employee_Satisfaction_Score'], kde=True) # specify the column to plot
plt.title('Distribution of Employee Satisfaction Scores') # set the title
plt.xlabel('Satisfaction Score') # set the x label
plt.ylabel('Count') # set the Y axis lebel
plt.show()
```

# Performance Score vs Monthly Salary

```python
# plot a scatter plot ( Monthy salary vs perfomance)
df.plot(kind='scatter', x='Performance_Score', y='Monthly_Salary', s=32, alpha=.8) 
plt.gca().spines[['top', 'right',]].set_visible(False)
```

It is evident that employees who had the highest perfomance scores also had higher monthly salary.

# Performance Score vs Monthly Salary by department

```python
# set the size of the figure
plt.figure(figsize=(10, 6))

# plot the scatter plot
sns.scatterplot(data=df, x='Performance_Score', y='Monthly_Salary', hue='Department')

# set the labels
plt.title('Performance Score vs Monthly Salary by Department')
plt.xlabel('Performance Score')
plt.ylabel('Monthly Salary')

# display the plot
plt.show() 
```

# Performance by department

```python
# compute performance score for each department
avg_performance_by_department = df.groupby('Department')['Performance_Score'].mean()
avg_performance_by_department
```

The Operations department had the highest perfomance score followed by the IT department.

# Age distribuion

```python

# Get and plot distribution of the age
sns.histplot(x='Age', data=df, kde=True)
plt.title('Employee Age Distribution') # specify the title
plt.show() # shw the plot 

```

All the age brackets have almost same amount of employees

# Education Level

```python
# plot education level
df.groupby('Education_Level').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

## Perfomance by education level

```python
# performance by education level
sns.catplot(x='Education_Level', hue='Performance_Score', kind='count', data=df)
plt.title('Perfomance by Education Level') # set the title
plt.show() # show the plot
```

Bachelor level employees have the highest performance scores (level 5) followed by high schoolers, masters and lastly PhD level employees. The distribution indicates that employees with a Bachelor's degree tend to perform better on average than those with a High School education.

## Distribution of Monthly Salary by Education level

```python
df[df['Monthly_Salary'] == df.Monthly_Salary.max()]['Education_Level'].hist(bins=25) # compute monthly salary for each education level

# labels for the plot
plt.xlabel('Educatin Level') # set X axis
plt.ylabel('Monthly Salary') # set Y axis
plt.title('Distribution of Monthly Salary by Education level') # set the title

plt.show() # display the plot
```

Bachelor level employees seems to be earning the highest amount of salary. Suprisingly, Masters and PhD holders have the lowest salaries indicating that this two groups of employees may be occupying positions that are not equated by high salaries as well as for the fact that there are not many employees that are masters and PhD holders.
