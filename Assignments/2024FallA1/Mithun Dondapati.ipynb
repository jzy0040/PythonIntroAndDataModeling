```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("D:/heart_cleveland.csv")
```

```python
# removing null vales
df = df.dropna()
df.isnull().sum()
```

```python
# Distribution of the 'age' column
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The age distribution plot helps us visualize the spread of ages among the patients in the dataset. 
The KDE line shows the probability density, indicating where most patients' ages are concentrated.

```python
# Distribution of 'chol' (cholesterol levels)
plt.figure(figsize=(10, 6))
sns.histplot(df['chol'], bins=20, kde=True)
plt.title('Cholesterol Level Distribution')
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.show()
```

This graph shows the distribution of cholesterol levels in the dataset. 
It helps identify common cholesterol levels and potential outliers, as well as general trends in the data.

```python
# Boxplot for 'trestbps' (resting blood pressure) by 'condition' (heart disease presence)
plt.figure(figsize=(10, 6))
sns.boxplot(x='condition', y='trestbps', data=df)
plt.title('Resting Blood Pressure by Heart Disease Condition')
plt.xlabel('Heart Disease Condition (0 = No, 1 = Yes)')
plt.ylabel('Resting Blood Pressure (trestbps)')
plt.show()
```

The boxplot compares resting blood pressure between patients with and without heart disease. 
It helps visualize differences in central tendencies and variability, highlighting whether resting blood pressure is higher for those with heart disease.

```python
# Univariate Graphical EDA - Histograms for key features
df[['age', 'chol', 'trestbps', 'thalach']].hist(bins=15, figsize=(10, 8), layout=(2, 2))
plt.suptitle('Distribution of Age, Cholesterol, Blood Pressure, and Max Heart Rate')
plt.show()
```

The histogram plot shows the distribution of four key features: age, cholesterol (chol), resting blood pressure (trestbps), and maximum heart rate achieved (thalach). Each subplot gives insights into the spread and central tendencies of these variables. The graphs highlight common values, potential skewness in the data, and outliers, helping to understand the general characteristics of the datasets numeric features.

```python
# Box plot to compare heart disease condition with cholesterol levels
plt.figure(figsize=(8, 6))
sns.boxplot(x='condition', y='chol', data=df)
plt.title('Cholesterol Levels by Heart Disease Condition')
plt.show()
```

This box plot compares cholesterol levels between patients with and without heart disease. The x-axis represents the heart disease condition (0 = no heart disease, 1 = heart disease), while the y-axis shows cholesterol levels. The plot highlights the median cholesterol levels and the spread (interquartile range) for both groups, with potential outliers indicated by points outside the whiskers. This visualization helps identify if cholesterol levels differ significantly between the two groups.

```python
# Scatter plots to explore relationships between features and the target variable
plt.figure(figsize=(10, 6))

# Age vs Max Heart Rate
plt.subplot(1, 2, 1)
sns.scatterplot(x='age', y='thalach', hue='condition', data=df)
plt.title('Age vs Max Heart Rate')
```

The scatter plot visualizes the relationship between age and maximum heart rate, with colors representing heart disease presence. It helps identify if certain age groups have lower or higher maximum heart rates and whether heart disease affects this relationship. This comparison may reveal trends, such as reduced heart rate with increasing age or disease.

```python
# Resting Blood Pressure vs Cholesterol
plt.subplot(1, 2, 2)
sns.scatterplot(x='trestbps', y='chol', hue='condition', data=df)
plt.title('Resting Blood Pressure vs Cholesterol')

plt.tight_layout()
plt.show()
```

This scatter plot explores the relationship between resting blood pressure (trestbps) and cholesterol (chol), with colors indicating the presence of heart disease. It shows how these two features vary together and whether higher values of blood pressure and cholesterol are more common among patients with heart disease. This plot helps identify potential patterns or correlations between these important cardiovascular metrics.

```python
# Correlation Matrix - Non-graphical EDA for multivariate relationships
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt='.2f')
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.show()
```

The correlation matrix visually represents the pairwise relationships between numeric features in the dataset. Each cell contains a correlation coefficient (ranging from -1 to 1), where positive values indicate a direct relationship and negative values indicate an inverse relationship between two features. Strong correlations (closer to -1 or 1) suggest that changes in one feature are closely linked to changes in another. This matrix helps identify key interactions between variables, aiding in feature selection and understanding dependencies that may affect heart disease prediction.

```python
# Pairplot for selected features
sns.pairplot(df[['age', 'trestbps', 'chol', 'thalach', 'condition']], hue='condition')
plt.show()
```
