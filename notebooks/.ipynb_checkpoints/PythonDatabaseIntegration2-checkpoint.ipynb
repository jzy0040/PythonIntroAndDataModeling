| `IS NULL` / `IS NOT NULL` | `email IS NOT NULL`            | Check empty values |

### Preparing the data

##### ✅ Reason #1: MySQL needs strict, predictable data types
Pandas and NumPy often store data in flexible formats (object, float64, datetime64) that don’t map directly to MySQL types.

If we don’t define compatible types in the CREATE TABLE statement, MySQL might reject rows or store them in less efficient formats (e.g., turning numbers into text).

Example:
1. Pandas float64 → MySQL DOUBLE
2. Pandas datetime64[ns] → MySQL DATETIME
3. Strings with >255 characters → MySQL TEXT (instead of VARCHAR(255))

##### ✅ Reason #2: Avoid insertion errors from NaN, NaT, and special values

Pandas uses NaN (float) and NaT (datetime) for missing values, but MySQL doesn’t recognize them — it expects NULL.

NumPy can also produce inf or -inf, which MySQL will reject unless explicitly handled.

All above efforts needs to be done before we can forward data to MySQL to ensure compatibility. Keep in mind, we are no longer just in Python environment. Instead, we are building a bridge between Python and MySQL. 

```python
### Data type conversion to make sure compatibility. 

# Try to parse any 'date'-like columns automatically
preview_cols = df.columns.tolist() # df is the dataframe. See the beginning of this project. 

# date format is a standard data type in Database.
# we need to handle this separately as following.
date_guess = [c for c in preview_cols if "date" in c.lower()]

# now we know which column is likely to be date format in database
# and it's stored in date_guess
# we can re-load the data
df = pd.read_csv(fname, parse_dates=date_guess or None)
```

```python
# column name can't have space as space is not recognized in database. 

# Clean column names: trim & replace spaces with underscores
df.columns = [str(c).strip().replace(" ", "_") for c in df.columns]
```

```python
# We don't want to manually handle the data type for the columns in the data
# We want something that can automatically deside the data type
# here is how to do so

# Python TYPE INFERENCE -> MYSQL TYPES

def mysql_type_from_series(s: pd.Series) -> str:
    if pd.api.types.is_datetime64_any_dtype(s):
        return "DATETIME"
    if pd.api.types.is_bool_dtype(s):
        return "TINYINT(1)"
    if pd.api.types.is_integer_dtype(s):
        # Pick INT vs BIGINT based on range
        minv, maxv = s.min(skipna=True), s.max(skipna=True)
        if pd.isna(minv) or pd.isna(maxv):
            return "BIGINT"
        fits_int = (-2147483648 <= int(minv)) and (int(maxv) <= 2147483647)
        return "INT" if fits_int else "BIGINT"
    if pd.api.types.is_float_dtype(s):
        return "DOUBLE"
    # strings/objects
    if s.notna().any():
        max_len = int(s.dropna().astype(str).map(len).max())
    else:
        max_len = 0
    return "TEXT" if max_len > 255 else "VARCHAR(255)"

col_types = {col: mysql_type_from_series(df[col]) for col in df.columns}
```

```python
# It is critical to set up the primary key for any database
# It is good to use what have already in the data
# however, if no column fit for the primary key, we can create one

# If there’s an 'id' column that looks integral, make it the primary key
primary_key = None
if "id" in df.columns and (pd.api.types.is_integer_dtype(df["id"]) or pd.api.types.is_float_dtype(df["id"])):
    primary_key = "id"
    if col_types["id"] not in ("INT", "BIGINT"):
        col_types["id"] = "BIGINT"
```

```python
# ========= BUILD CREATE TABLE =========

def q(name: str) -> str:
    # this is utility function to quote SQL identifier,
    # which works for take names and columns names
    return f"`{name}`"

columns_sql = [f"{q(c)} {t}" for c, t in col_types.items()]
if primary_key:
    columns_sql.append(f"PRIMARY KEY ({q(primary_key)})")

create_sql = (
    f"CREATE TABLE IF NOT EXISTS {q(TABLE_NAME)} (\n  "
    + ",\n  ".join(columns_sql)
    + "\n) ENGINE=InnoDB DEFAULT CHARSET=utf8mb4;"
)
```

```python
# ========= INSERT DATA (chunked) =========

# we handle NaN/NaT values while inserting the records to the database. 
def to_python(value):
    """Convert pandas/NaN/NaT to Python types mysql.connector accepts."""
    if pd.isna(value):
        return None
    if isinstance(value, (pd.Timestamp, datetime)):
        return value.to_pydatetime() if hasattr(value, "to_pydatetime") else value
    try:
        if isinstance(value, np.integer):
            return int(value)
        if isinstance(value, np.floating):
            v = float(value)
            if math.isinf(v) or math.isnan(v):
                return None
            return v
        if isinstance(value, np.bool_):
            return bool(value)
    except Exception:
        pass
    return value
```

```python
# connect to database using MYSQL_CONFIG
conn = mysql.connector.connect(**MYSQL_CONFIG)

# create the cursor
cur = conn.cursor()

# Create table
cur.execute(create_sql)

# Prepare INSERT
cols = list(df.columns)
placeholders = ", ".join(["%s"] * len(cols))

# insert multiple data using for loop
insert_sql = f"INSERT INTO {q(TABLE_NAME)} ({', '.join(q(c) for c in cols)}) VALUES ({placeholders})"

# Insert in batches

# BATCH_SIZE controls how many rows are inserted into
# the database at once during each executemany() call.
# This is because 1. inserting thousands of rows one by one is very slow
# 2. Loading millions of rows at once into a single insert can cause 
# memory errors or overload the database.
# 3. Each batch sends one request containing multiple rows, 
# reducing the number of network calls.

BATCH_SIZE = 200 
rows_inserted = 0
for start in range(0, len(df), BATCH_SIZE):
    chunk = df.iloc[start:start + BATCH_SIZE]
    data = [tuple(to_python(v) for v in row) for row in chunk.itertuples(index=False, name=None)]
    if data:
        cur.executemany(insert_sql, data)
        conn.commit()
        rows_inserted += len(data)

cur.close()
conn.close()
```

# Part 2 Data Retrieve Using Conditions

```python
### Retrieve data using condition statement

# ==== Connection settings ====
MYSQL_CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": getpass("Enter MySQL password: "),
    "database": "HouseDatabase",
    "port": 3306
}
```

```python
# Please pick one query and comment out the others in this cell to proceed with the data retrieve. 
# By default, it's the query with the WHERE clause. 

# ==== SQL query with WHERE condition ====
# We want Houses in AUBURN with price under 800,000
query = """
SELECT *
FROM house
WHERE Subdivision = 'AUBURN'
  AND Price < 800000;
"""

# # ==== SQL query using LIKE ====
# # Find all properties where the Subdivision name starts with "The".
# # % = any number of characters, _ = one character
# query = """
# SELECT *
# FROM house
# WHERE Subdivision LIKE 'The%';
# """

# # ==== SQL query using BETWEEN ====
# # Retrieve properties priced between 500,000 and 1,000,000.
# # BETWEEN is inclusive: includes both lower and upper bounds
# query = """
# SELECT *
# FROM house
# WHERE Price BETWEEN 500000 AND 1000000;
# """

# # ==== SQL query using IN ====
# # Get all houses in "EAST LAKE", "GROVE HILL", or "AUBURN".
# query = """
# SELECT *
# FROM house
# WHERE Subdivision IN ('EAST LAKE', 'GROVE HILL', 'AUBURN');
# """
```

```python
# ==== Connect and execute ====
conn = mysql.connector.connect(**MYSQL_CONFIG)
cursor = conn.cursor()

# Execute query
cursor.execute(query)

# Fetch results and column names
rows = cursor.fetchall()
columns = [desc[0] for desc in cursor.description]

# Convert to DataFrame for easier viewing
df_retrieved = pd.DataFrame(rows, columns=columns)

# Close connection
cursor.close()
conn.close()

# Show result
print(df_retrieved.head())
```

```python
import mysql.connector as mysql
import pandas as pd

# ==== Connection settings ====
MYSQL_CONFIG = {
    "host": "localhost",
    "user": "root",
    "password": getpass("Enter MySQL password: "),
    "database": "HouseDatabase",
    "port": 3306
}

# ==== Your query (example: houses priced > 1M) ====
query = """
SELECT *
FROM house
WHERE Price > 1000000;
"""

# ==== Connect and retrieve ====
conn = mysql.connect(**MYSQL_CONFIG)
df = pd.read_sql(query, conn)
conn.close()

print(df.head())
```
