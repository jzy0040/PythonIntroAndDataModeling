```python
### We need to load the data first. 

import pandas as pd

gradebook = pd.read_csv('../../data/FullGrade.csv')

### let's preview the gradebook
gradebook
```

```python
### Function to calculate the total grade
def TotalGrade(gradebook):
    """
    Calculate total grade by sum up all the grade items for all students

    Args:
        gradebook(dataframe):  a dataframe contains all grade for each grade item

    Return:
        gradebook(dataframe): a new dataframe by adding the total grade as the last column to the input gradeboook.
    """
    gradebook['TotalGrade'] = gradebook.iloc[:, 1:].sum(axis = 1)
    return gradebook

TotalGrade(gradebook)
```

```python
### Calculating the latter grade based on the grading scale
def LetterGrade(TotalScore):
    """
    Assign letter grade to one student based on the total numerical grade.

    Args:
        TotalScore(Float): one student's total grade(out of 1000)

    Return:
        letter_grade(String): The corresponding letter grade. 
    """
    percentage = TotalScore/10

    if percentage >= 93:
        return 'A'
    elif 90 <= percentage <= 92.99:
        return 'A-'
    elif 87 <= percentage <= 89.99:
        return 'B+'
    elif 83 <= percentage <= 86.99:
        return 'B'
    elif 80 <= percentage <= 82.99:
        return 'B-'
    elif 77 <= percentage <= 79.99:
        return 'C+'
    elif 73 <= percentage <=76.99:
        return 'C'
    elif 70 <= percentage <= 72.99:
        return 'C-'
    elif 67 <= percentage <= 69.99:
        return 'D+'
    elif 63 <= percentage <= 66.99:
        return 'D'
    elif 60 <= percentage <= 62.99:
        return 'D-'
    else:
        return 'F'

def LetterGradeToClass(gradebook):
    """
    Calculate letter grade to all students

    Args:
        gradebook(dataframe): all students' grade(out of 1000)

    Return:
        gradebook(dataframe): Adding the letter grade as the last column for all students. 
    """
    all_letter_grade = [] ### a temp place holder for store the letter grade
    for grade in gradebook['TotalGrade']:
        all_letter_grade.append(LetterGrade(grade))
    
    gradebook['LetterGrade'] = pd.Series(all_letter_grade)
    return gradebook
```

```python
LetterGradeToClass(gradebook)
```

```python
### Passing rate calculation
def PassingRate(gradebook):
    """
    Calculate passing rate
    TotalGrade >= 73

    Args:
        gradebook(dataframe): all students' grade(out of 1000)

    Return:
        gradebook(dataframe): add pass or not as the last column to the gradebook
        rate(float): a percentage value representing the passing rate
    """
    passFail = [] ### Yes or No value to indicate pass or not
    num_passed = 0

    for grade in gradebook['TotalGrade']:
        if grade >= 730:
            passFail.append('Yes')
            num_passed = num_passed + 1
        else:
            passFail.append('No')
    gradebook['Pass'] = pd.Series(passFail)
    rate = num_passed / len(passFail)
    return gradebook, rate

PassingRate(gradebook)
```

```python
def averageGrade(gradebook):
    """
    Calculate average performance

    Args:
        gradebook(dataframe): all students' grade(out of 1000)

    Return:
        averageGrade(Float): average grade
        averageLetterGrade(String): average letter grade
    """

    averageGrade = gradebook['TotalGrade'].mean()
    averageLetterGrade = LetterGrade(gradebook['TotalGrade'].mean())
    return averageGrade, averageLetterGrade
averageGrade(gradebook)
```
