# Preattentive Attributes

It's the visual properties that are processed by the brain very rapidly and without conscious effort. These attributes can help in guiding attention and highlighting important information in visualizations. Some common preattentive attributes include:

<span style='color:blue'>Color</span>: Differences in color can be perceived quickly, making it useful for encoding categories or highlighting specific data points.

<span style='color:blue'>Size</span>: Differences in size can also be perceived rapidly, helping to indicate magnitude or importance.

<span style='color:blue'>Shape</span>: Different shapes can be used to encode categories or distinguish between different types of data.

<span style='color:blue'>Orientation</span>: Changes in orientation, such as slope or angle, can be detected quickly and used to convey information.

<span style='color:blue'>Movement</span>: Dynamic changes or motion can attract attention and convey changes over time.

<span style='color:blue'>Spatial Position</span>: Differences in position along a common scale or axis can be perceived rapidly and used to compare values.

These examples showcase how preattentive attributes can be used to encode data in visualizations to make patterns and trends more easily perceptible.

Risk-Seeking: Risk-seeking individuals prefer uncertain outcomes with higher variability, even if the expected value (average outcome) is the same as a certain outcome. They are willing to take on more risk in exchange for the possibility of higher rewards. Risk-seeking behavior is typically associated with individuals who enjoy taking risks or who have a high tolerance for uncertainty.

Risk-Neutral: Risk-neutral individuals are indifferent between uncertain outcomes and certain outcomes with the same expected value. They make decisions based solely on maximizing expected utility or value, without considering the variability or riskiness of the outcomes. Risk-neutral behavior is often assumed in economic models as a simplifying assumption.

Risk-Averse: Risk-averse individuals prefer certain outcomes over uncertain outcomes, even if the expected value of the uncertain outcome is the same or higher. They are willing to accept lower expected returns to reduce the variability or risk of their outcomes. Risk-averse behavior is common among individuals who are averse to uncertainty or who prioritize stability and security in their decision-making.

### Position and Shape

Box plots can effectively convey information and demonstrate the concept of preattentive attributes, particularly focusing on position and shape.

A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It is used to visualize the spread and skewness of the data along with identifying outlier. For reference, https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

<span style='color:blue'>Position</span>: The central line within the box represents the median of the data, while the top and bottom edges of the box represent the upper and lower quartiles, respectively. The position of these elements along the y-axis helps viewers quickly assess the central tendency and spread of the data.

<span style='color:blue'>Shape</span>: The box plot itself is a simple shape, consisting of a rectangular box with lines (whiskers) extending from the top and bottom edges. The length of the box and whiskers, as well as any outliers or extreme values, provide additional information about the distribution of the data.

While box plots may not use color or size to the same extent as other visualizations, their effectiveness lies in their ability to efficiently communicate important summary statistics using preattentive attributes like position and shape.

### Movement

By nature, a lot of creatures are more sensitive to moving objects. For example, dog and cat like to chase some running targets. Frogs have a visual system that is adapted to see moving objects. They are nearsighted and can only focus on objects that are passing by or flying within six inches. Frogs also find it hard to see still prey. Although human's visual system are less stronger than those animals towards the moving objects. But our eyes are also quite sensitive to moving objects. By nature, we trends to learn information from moving objects.

```python
import pygame
import random

# Initialize pygame
pygame.init()

# Game Constants
WIDTH, HEIGHT = 500, 400
BALL_RADIUS = 15
PADDLE_WIDTH, PADDLE_HEIGHT = 100, 10
PADDLE_SPEED = 6
MAX_BALLS = 3  # Maximum number of balls allowed
MAX_BALLS_SPEED = 5  # Maximum number of balls allowed

# Colors
BLACK = (0, 0, 0)
WHITE = (255, 255, 255)
COLORS = [(255, 0, 0), (0, 255, 0), (0, 0, 255)]  # Colors for balls

# Get user input
num_balls = min(max(int(input("Enter number of balls (1-3): ")), 1), MAX_BALLS)
ball_speed = min(max(int(input("Enter ball speed (1-5): ")), 1), MAX_BALLS_SPEED)

# Set up display
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Bouncing Ball Game")

# Ball positions and velocities
balls = []
for i in range(num_balls):
    ball_x = random.randint(50, WIDTH - 50)
    ball_y = random.randint(50, HEIGHT // 2)
    speed_x = random.choice([-1, 1]) * ball_speed
    speed_y = -ball_speed
    balls.append({'x': ball_x, 'y': ball_y, 'speed_x': speed_x, 'speed_y': speed_y, 'color': COLORS[i]})

# Paddle position
paddle_x = (WIDTH - PADDLE_WIDTH) // 2
paddle_y = HEIGHT - 30  # Near the bottom

# Game loop
running = True
clock = pygame.time.Clock()

# Start the timer
start_time = pygame.time.get_ticks()

while running:
    screen.fill(BLACK)  # Clear screen

    # Calculate elapsed time
    elapsed_time = (pygame.time.get_ticks() - start_time) // 1000  # Convert milliseconds to seconds

    # Draw paddle
    pygame.draw.rect(screen, WHITE, (paddle_x, paddle_y, PADDLE_WIDTH, PADDLE_HEIGHT))

    # Handle user input (paddle movement)
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and paddle_x > 0:
        paddle_x -= PADDLE_SPEED
    if keys[pygame.K_RIGHT] and paddle_x < WIDTH - PADDLE_WIDTH:
        paddle_x += PADDLE_SPEED

    # Update and draw each ball
    for ball in balls:
        # Move ball
        ball['x'] += ball['speed_x']
        ball['y'] += ball['speed_y']

        # Ball collision with walls
        if ball['x'] - BALL_RADIUS < 0 or ball['x'] + BALL_RADIUS > WIDTH:
            ball['speed_x'] = -ball['speed_x']
        if ball['y'] - BALL_RADIUS < 0:
            ball['speed_y'] = -ball['speed_y']

        # Ball collision with paddle
        if (paddle_y < ball['y'] + BALL_RADIUS < paddle_y + PADDLE_HEIGHT) and (paddle_x < ball['x'] < paddle_x + PADDLE_WIDTH):
            ball['speed_y'] = -ball['speed_y']

        # Ball falls off bottom (Game Over)
        if ball['y'] + BALL_RADIUS > HEIGHT:
            print("Game Over!")
            running = False  # End game

        # Draw the ball
        pygame.draw.circle(screen, ball['color'], (int(ball['x']), int(ball['y'])), BALL_RADIUS)

    # Display elapsed time
    font = pygame.font.Font(None, 36)
    time_text = font.render(f"Time: {elapsed_time} sec", True, WHITE)
    screen.blit(time_text, (10, 10))

    # Refresh display
    pygame.display.flip()
    clock.tick(60)  # 60 FPS

    # Handle quit event
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
```

### Spatial position

Spatial position in data visualization refers to graphical representation of geographic or positional data in two or three dimensions. As a preattentive attribute, it refers to the ability to quickly perceive differences in the position of visual elements. In data visualization, spatial position can be used to encode information, such as differentiating between categories or showing trends in the data.


```python
import seaborn as sns

# Generate random spatial data
x = np.random.uniform(0, 100, 1000)
y = np.random.uniform(0, 100, 1000)

# Create a heatmap
sns.kdeplot(x=x, y=y, cmap="Blues", fill=True)
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')
plt.title('Spatial Density Plot')
plt.show()
```

Spatial position is commonly used in map visualization. The latitude, longitude, and sometimes altitude define the location of objects on a map.
