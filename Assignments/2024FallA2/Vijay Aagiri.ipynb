# **Test of Significance(Z-Test):**

**Objective:**
In this problem, I wanted to find out if the new training program significantly reduced the average assembly time for employees.

**Description:**
The average assembly time before training was 35 minutes. After the training, I collected data from a sample of 40 employees, and their average assembly time came out to be 33 minutes. The standard deviation for the population is 5 minutes. My goal was to determine whether this decrease was statistically significant or just a random fluctuation.

**Steps I Took:**
- I set up my hypotheses:
  - Null Hypothesis (\( H_0 \)): \( \mu \geq 35 \)
  - Alternative Hypothesis (\( H_1 \)): \( \mu < 35 \)
- I calculated the standard error using the population standard deviation and the sample size.
- Then, I computed the Z-statistic using the sample mean, population mean, and standard error.
- Next, I found the critical value for a one-tailed test at a significance level of 5%.
- I compared the Z-statistic with the critical value and calculated the p-value.
- Based on this, I made my decision. If the Z-statistic was smaller than the critical value, I would reject the null hypothesis.


```python
import scipy.stats as stats
import numpy as np

# I have the given data for Question 1
population_mean = 35  # This is the average time before training
sample_mean = 33  # This is the average time after training
population_std_dev = 5  # This is the standard deviation of assembly times
sample_size = 40  # This is the number of employees in the sample
alpha = 0.05  # I choose a 5% significance level

# I calculated the standard error
standard_error = population_std_dev / np.sqrt(sample_size)

# I found the Z-statistic using the given formula
z_statistic = (sample_mean - population_mean) / standard_error

# I got the critical value for a one-tailed test
critical_value = stats.norm.ppf(alpha)

# I calculated the p-value based on the Z-statistic
p_value = stats.norm.cdf(z_statistic)

# I printed the Z-statistic, critical value, and p-value
print(f"Z-statistic: {z_statistic:.2f}")
print(f"Critical value: {critical_value:.2f}")
print(f"P-value: {p_value:.4f}")

# I decided whether to reject or not reject the null hypothesis
if z_statistic < critical_value:
    print("Reject the null hypothesis: The training reduces assembly time.")
else:
    print("Fail to reject the null hypothesis.")
```

I did a Z-test to see if the new training reduced assembly time. The Z-statistic was -2.53, which is lower than the critical value of -1.64. This shows a significant drop in time. The p-value of 0.0057, being less than 0.05, also supports this. So, I rejected the null hypothesis and concluded that the training program successfully reduced the average assembly time.

# **Test of Significance(T-Test):**

**Objective:**
In this problem, I wanted to find out if graduate students study more than 25 hours per week on average.

**Description:**
I collected data from a sample of 15 graduate students, and their average study time was 27 hours per week. The sample standard deviation was 4.5 hours. My goal was to determine whether this average study time is significantly higher than 25 hours per week.

**Steps I Took:**
- I set up my hypotheses:
  - Null Hypothesis (\( H_0 \)): \( \mu \leq 25 \)
  - Alternative Hypothesis (\( H_1 \)): \( \mu > 25 \)
- I calculated the standard error using the sample standard deviation and the sample size.
- Then, I computed the T-statistic using the sample mean, population mean, and standard error.
- Next, I found the critical T-value for a one-tailed test with 14 degrees of freedom (sample size minus 1) at a significance level of 5%.
- I compared the T-statistic with the critical T-value and calculated the p-value.
- Finally, I made my decision. If the T-statistic was greater than the critical T-value, I would reject the null hypothesis.

```python
# I have the given data for Question 2
sample_mean_study = 27  # This is the average study time for students
population_mean_study = 25  # This is the target average study time
sample_std_dev_study = 4.5  # This is the standard deviation for the sample
sample_size_study = 15  # This is the number of students in the sample
alpha_study = 0.05  # I choose a 5% significance level

# I calculated the standard error for the T-test
standard_error_study = sample_std_dev_study / np.sqrt(sample_size_study)

# I found the T-statistic using the given formula
t_statistic = (sample_mean_study - population_mean_study) / standard_error_study

# I got the critical T-value for a one-tailed test
degrees_of_freedom = sample_size_study - 1
critical_t_value = stats.t.ppf(1 - alpha_study, df=degrees_of_freedom)

# I calculated the p-value based on the T-statistic
p_value_study = 1 - stats.t.cdf(t_statistic, df=degrees_of_freedom)

# I printed the T-statistic, critical T-value, and p-value
print(f"T-statistic: {t_statistic:.2f}")
print(f"Critical t-value: {critical_t_value:.2f}")
print(f"P-value: {p_value_study:.4f}")

# I decided whether to reject or not reject the null hypothesis
if t_statistic > critical_t_value:
    print("Reject the null hypothesis: Students study more than 25 hours per week.")
else:
    print("Fail to reject the null hypothesis.")
```

I did a T-test to see if students study more than 25 hours per week. The T-statistic was 1.72, which is slightly lower than the critical value of 1.76. The p-value was 0.0536, which is greater than 0.05. Since the T-statistic didn’t exceed the critical value and the p-value was higher than 0.05, I failed to reject the null hypothesis. This means there isn’t enough evidence to say that students study more than 25 hours per week.

# **Regression Analysis:**

**Objective:**
The objective was to fit a simple linear regression model to analyze the relationship between hours of study and exam scores.

**Description:**
I had data on the hours of study and corresponding exam scores for 30 students. My goal was to see if there was a relationship between these two variables. I used the hours of study as the independent variable (X) and the exam scores as the dependent variable (Y).

**Steps I Took:**
- I organized the data into a DataFrame with columns for Hours of Study and Scores.
- I fit a linear regression model using Hours of Study as the independent variable and Scores as the dependent variable.
- I obtained the coefficients (intercept and slope) to describe the equation of the regression line.
- I plotted a scatter plot with a regression line to visualize the relationship between the variables.

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import statsmodels.api as sm

# I created a dictionary with data for Regression Analysis
data = {
    'Hours_Study': [0, 1, 2, 3, 4, 5, 6, 7, 8],  # Hours of study (independent variable)
    'Score': [0, 10, 20, 30, 40, 50, 60, 70, 80]  # Exam scores (dependent variable)
}

# I converted the data into a DataFrame
df = pd.DataFrame(data)

# I set up X and y variables for the regression model
X = df[['Hours_Study']]  # Independent variable
y = df['Score']  # Dependent variable

# I created and trained the regression model
model = LinearRegression()
model.fit(X, y)

# I got the intercept and slope from the model
intercept = model.intercept_
slope = model.coef_[0]

# I printed the regression equation
print(f"Regression Equation: Exam Score = {intercept:.2f} + {slope:.2f} * Hours_Study")

# I plotted the data points and the regression line
plt.scatter(X, y, color='blue', label='Data Points')  # I plotted the points
plt.plot(X, model.predict(X), color='red', label='Regression Line')  # I plotted the line
plt.title('Hours of Study vs. Exam Score')
plt.xlabel('Hours of Study')
plt.ylabel('Exam Score')
plt.legend()
plt.show()

# I got the predictions and calculated residuals
y_pred = model.predict(X)
residuals = y - y_pred

# I plotted residuals vs. fitted values to check for homoscedasticity
plt.scatter(y_pred, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Residuals vs. Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.show()

# I plotted a histogram of residuals to check for normality
sns.histplot(residuals, kde=True, color='orange')
plt.title('Histogram of Residuals')
plt.xlabel('Residuals')
plt.show()

# I plotted a Q-Q plot to check for normality of residuals
sm.qqplot(residuals, line='45')
plt.title('Q-Q Plot of Residuals')
plt.show()

# I calculated R-squared to check model performance
r_squared = model.score(X, y)
print(f"R-squared: {r_squared:.2f}")
```

The regression line plot shows a perfect linear relationship between study hours and exam scores. The residuals vs. fitted values plot indicates that most residuals are close to zero, with some larger ones at the extremes. The histogram suggests balanced errors, but the shape is not perfectly normal. The Q-Q plot shows that residuals are mostly normal, with slight deviations at the ends.

# **Linear Optimization:**

**Objective:**
In this problem, I wanted to maximize the production of three products (A, B, and C) while considering constraints on Machine X and Machine Y usage.

**Description:**
Each product (A, B, and C) requires different amounts of Machine X and Y hours. Machine X has a total of 100 hours available, and Machine Y has 85 hours available. The goal was to find the optimal number of units for each product to maximize production.

**Steps I Took:**
- I defined the objective function to maximize total production: \( A + B + C \).
- I set the constraints for Machine X and Machine Y:
  - Machine X: \( 2A + 1B + 3C \leq 100 \)
  - Machine Y: \( 4A + 3B + 2C \leq 85 \)
- I specified the non-negativity bounds for each product (units can't be negative).
- I used linear programming to find the optimal solution within the constraints.
- I checked the results to determine the optimal number of units for each product.

```python
from scipy.optimize import linprog

# I set the coefficients for the objective function (I want to maximize A + B + C)
c = [-1, -1, -1]  # I use negatives for maximization in linprog

# I set the constraints matrix (hours required for Machine X and Y)
A = [
    [2, 1, 3],  # Hours per unit for Machine X
    [4, 3, 2]   # Hours per unit for Machine Y
]

# I set the maximum available hours for Machine X and Machine Y
b = [100, 85]

# I set the bounds for each variable (cannot be negative)
x_bounds = (0, None)
bounds = [x_bounds, x_bounds, x_bounds]

# I solve the linear programming problem using linprog with the 'highs' method
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

# I print the optimal solution if found
if result.success:
    print("Optimal Solution Found:")
    print(f"Units of Product A: {result.x[0]:.2f}")
    print(f"Units of Product B: {result.x[1]:.2f}")
    print(f"Units of Product C: {result.x[2]:.2f}")
else:
    print("Optimization Failed!")
```

The results suggest producing 0 units of Product A, about 7.86 units of Product B, and 30.71 units of Product C to maximize production. This combination works best given the current resource constraints.

Machine Y is the limiting factor as it reaches its full capacity of 85 hours, while Machine X uses nearly all of its 100 hours.

To boost production, upgrading Machine Y would be the best option, as it would provide more flexibility to produce more units of Products B and C and potentially add Product A in the future.
