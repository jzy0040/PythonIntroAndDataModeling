```python
# So let's get to the main part of the assignment where we decided some specifications
# Need a wine of quality 7 and above.
# Considering, I am acidic so need a balanced range for fixed and volatile acidity.
# I am also diabetic so wine should have low residual sugar
# I need to spend quality time at party hanging out with friends so need to select a wine with a moderate alcohol level.
# Last but not least lately i have been trying to have a healthy diet so need a wine with low free and total sulfur dioxide and also a balance pH.
```

```python
# Need a wine of quality 7 and above.
quality_wine = wine[wine['quality'] >= 7]
quality_wine
#quality_wine is the new dataset with then wine quality greater than or equal to 7
```

```python
quality_wine.shape
```

```python
# Considering, I am acidic so need a balanced range for fixed and volatile acidity.
#lets see the plot for Fixed Acidity vs Quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='fixed acidity', data=quality_wine)
plt.title('Fixed Acidity vs Quality')
plt.show()
#lets see the plot for Volatile Acidity vs Quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='volatile acidity', data=quality_wine)
plt.title('volatile acidity vs Quality')
plt.show()
```

```python
#As we see both of the box plots lets consider the range which have more options and also low value. 
# So for fixed acidity let choose range (6.2,7.5) and 
#for volatile acidity lets choose a range (0.2,0.35) 
```

```python
quality_wine1 = quality_wine[(quality_wine['volatile acidity'] >= 0.2) & (quality_wine['volatile acidity'] <= 0.35)&(quality_wine['fixed acidity'] >= 6.25) & (quality_wine['fixed acidity'] <= 7.5)]
quality_wine1
```

```python
#acidity is good now lets get to residual sugar we need to be low.
#lets see box plot of Residual Sugar vs Quality
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='residual sugar', data=quality_wine1)
plt.title('Residual Sugar vs Quality')
plt.show()
```

```python
#coming to recidual sugar quality 9 is low but it doesn't have many options. So we will search the wines with residual sugar range as (2.5,5.0)
quality_wine2 = quality_wine1[(quality_wine1['residual sugar'] >= 2.5) & (quality_wine1['residual sugar'] <= 5.0)]
quality_wine2
```

```python
quality_wine2.shape
```

```python
quality_wine2.describe()
```

```python
# now lets come to the alcohol content part we need it be not too low or not high we need it to be moderate
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='alcohol', data=quality_wine2)
plt.title('Alcohol vs Quality')
plt.show()
```

```python
# the plot shows we can have multiple options for the values between 11.3 and 12.5(our criteria is also to have moderate alcohol level)
#lets select the wine which have alcohol in range (11.3,12.5)
quality_wine3 = quality_wine2[(quality_wine2['alcohol'] >= 11.3) & (quality_wine2['alcohol'] <= 12.5)]
quality_wine3
```

```python
quality_wine3.describe()
```

```python
# quality, acidity,residual sugar, and alcohol are sorted so let's see some plots of free sulfur dioxide vs Quality and total sulfur dioxide vs Quality
# low levels of free sulfur dioxide and total sulfur dioxide are best for health
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='free sulfur dioxide', data=quality_wine3)
plt.title('free sulfur dioxide vs Quality')
plt.show()
#
plt.figure(figsize=(10, 6))
sns.boxplot(x='quality', y='total sulfur dioxide', data=quality_wine3)
plt.title('total sulfur dioxide vs Quality')
plt.show()
```

```python
# after seeing the plots lets consider the range of (min,25%) for both free sulfur dioxide and total sulfur dioxide
quality_wine4 = quality_wine3[(quality_wine3['free sulfur dioxide'] >= 10.000000) & (quality_wine3['free sulfur dioxide'] <= 26.000000)&(quality_wine3['total sulfur dioxide'] >= 37.000000) & (quality_wine3['total sulfur dioxide'] <= 99.000000)]
quality_wine4
```

```python
#quality_wine4 is our desired wine option which checks all are specifications
```

```python
#lets just get a glance at heat map where we can see correlations between all the factors of wine quality dataset
# lets take the dataset where quality is more than or equal to 7
plt.figure(figsize=(10, 8))
correlation = quality_wine1[['fixed acidity', 'volatile acidity', 'residual sugar', 'alcohol', 'free sulfur dioxide', 'total sulfur dioxide', 'pH', 'quality']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap of Key Features')
plt.show()
#In heat map cool colors represent the low value, whereas warm colors are for high values
```
