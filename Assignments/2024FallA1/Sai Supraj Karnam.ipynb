```python
import pandas as pd

# Load the CSV file
file_path = '/Users/Supraj/Downloads/FIFA17_official_data.CSV'
data = pd.read_csv(file_path)

# displaying the first few Rows of the Data Frame
print(data.head())
```

```python
# displaying the info of the dataset
print(data.info())
```

```python
# Summary statistics for numeric columns
print(data.describe())
```

```python
import matplotlib.pyplot as plt

# Plot a histogram for Age
data['Age'].hist(bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selecting only numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# orrelation matrix
corr_matrix = numeric_data.corr()

# heatmap plotting
plt.figure(figsize=(15, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Scatter plot for Age vs Overall rating
plt.scatter(data['Age'], data['Overall'])
plt.title('Age vs Overall Rating')
plt.xlabel('Age')
plt.ylabel('Overall Rating')
plt.show()
```
