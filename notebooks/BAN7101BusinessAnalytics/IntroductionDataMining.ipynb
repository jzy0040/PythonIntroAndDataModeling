### Black Box Model

The model operates based on inputs and outputs without revealing their internal workings. Black Box models focus on the performance or accuracy of the models rather than understanding how they arrive at their decisions. In this sense, a black box model is one where the logic or algorithm used to make predictions or decisions is not transparent or easily interpretable.

While black box models can often achieve high levels of accuracy, they can be challenging to interpret, which can be a significant drawback in certain applications where transparency and interpretability are important.

In recent years, there has been a growing interest in techniques and methods to make black box models more interpretable, such as using model-agnostic methods like SHAP (SHapley Additive exPlanations) values or LIME (Local Interpretable Model-agnostic Explanations) to provide explanations for individual predictions.

All machine learning and Artificial Intelligence topics we cover in this semester fall into the category of Black Box Model.

### Clustering 

Clustering is an unsupervised machine learning technique in machine learning. It groups a set of objects in such a way that objects in the same group are more similar to each other than to those in other groups.

```python
### this is an example of cluster algorithm
### we will cover this topic later this semester
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering

def loadDataSet():
    a = []
    f = open("../../data/B.txt",'r')
    lines = f.readlines()   #make lines a list, has index inside already
    f.close()
    for line in lines:
        b = ['', '']
        b[0],b[1] = line.split()
        a.append(b)
    dataSet = np.array(a,dtype=float)   
    return dataSet

X = loadDataSet()

cluster = AgglomerativeClustering(n_clusters = 2, linkage = 'single')  
cluster.fit_predict(X)  
print(cluster.labels_)  
print("Figure below is the clustered result.")

labels = range(0, 2)  
plt.figure(figsize = (5, 5))  
plt.subplots_adjust(bottom = 0.1)  
plt.scatter(X[:, 0], X[:, 1], c = cluster.labels_, cmap ='rainbow')
plt.show()
```

### Anomaly Detection

The identification of rare events, items, or observations which are suspicious because they differ significantly from standard behaviors or patterns.

Anomaly detection is a significant problem faced in several research areas. Detection and correctly classification of unseen data as anomalous is a challenging issue that has been handled in many different ways over the years.

<span style='color:blue'>Example Paper:</span> GANomaly: Semi-Supervised Anomaly Detection via Adversarial Training(https://arxiv.org/abs/1805.06725)

Some common Anomaly Detection tasks include:

Statistical Methods: Statistical methods such as Z-score, Grubbs' test, and Dixon's Q test can be used to detect anomalies based on the statistical properties of the data.

Machine Learning Algorithms: Machine learning algorithms such as Isolation Forest, One-Class SVM, and k-means clustering can be trained on normal data and then used to detect anomalies based on deviations from the learned patterns.

Time Series Analysis: For time-series data, techniques such as Seasonal-Trend decomposition using LOESS (STL) or AutoRegressive Integrated Moving Average (ARIMA) models can be used to detect anomalies in temporal patterns.

Deep Learning: Deep learning techniques, especially autoencoders, have been used for anomaly detection by reconstructing input data and identifying instances with high reconstruction errors. For example, GAN based anomaly detection was a very popular topic in around 2017. 

Unsupervised Learning: Unsupervised learning techniques such as density-based clustering (e.g., DBSCAN) can be used to identify dense regions of data and flag data points that fall outside these regions as anomalies.

Ensemble Methods: Ensemble methods such as Random Cut Forests (RCF) combine multiple anomaly detection algorithms to improve overall performance and robustness.

### Text Mining

Text mining, also known as text analytics, is the process of deriving high-quality information from text. It involves extracting valuable insights, patterns, and knowledge from unstructured text data. Text mining techniques can be used to analyze large volumes of text data efficiently and effectively. 

Some common text mining tasks include:

1. Text Classification: Classifying text documents into predefined categories or labels based on their content. For example, classifying emails as spam or non-spam, or classifying news articles by topic.

2. Sentiment Analysis: Determining the sentiment or opinion expressed in a piece of text. Sentiment analysis can be used to analyze customer reviews, social media posts, and other text data to understand customer opinions and attitudes.

3. Named Entity Recognition (NER): Identifying and classifying named entities such as names of people, organizations, locations, dates, and numerical expressions in text. NER is commonly used in information retrieval and text mining applications.

4. Topic Modeling: Identifying topics or themes present in a collection of text documents. Topic modeling algorithms, such as Latent Dirichlet Allocation (LDA), can be used to uncover hidden thematic structures in text data.

5. Text Summarization: Automatically generating a concise summary of a longer piece of text. Text summarization techniques can be used to extract key information from large volumes of text data.

6. Entity Sentiment Analysis: Analyzing the sentiment associated with specific entities mentioned in text. This can provide deeper insights into the opinions and attitudes expressed towards different entities.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax

# tweet = "@MehranShakarami today's cold @ home "
# tweet = 'Great content! subscribed '

tweet = 'Bad product, do not buy it'

# precprcess tweet
tweet_words = []

for word in tweet.split(' '):
    if word.startswith('@') and len(word) > 1:
        word = '@user'
    
    elif word.startswith('http'):
        word = "http"
    tweet_words.append(word)

tweet_proc = " ".join(tweet_words)

# load model and tokenizer
roberta = "cardiffnlp/twitter-roberta-base-sentiment"

model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)

labels = ['Negative', 'Neutral', 'Positive']

# sentiment analysis
encoded_tweet = tokenizer(tweet_proc, return_tensors='pt')
# output = model(encoded_tweet['input_ids'], encoded_tweet['attention_mask'])
output = model(**encoded_tweet)

scores = output[0][0].detach().numpy()
scores = softmax(scores)

for i in range(len(scores)):
    
    l = labels[i]
    s = scores[i]
    print(l,s)
```
