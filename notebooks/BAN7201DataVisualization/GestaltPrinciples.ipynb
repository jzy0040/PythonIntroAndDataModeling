# Gestalt Principles

Gestalt is a German word indicates shapes or forms and means “unified whole” in English. <br>
Gestalt psychology is a branch of psychology that focuses on the organization and interpretation of sensory information to represent and comprehend complex information.<br>
In general, this principle will help your viz more clear and effective.

There are varying interpretations of the exact meaning of Gestalt principles. Try to be flexible when design your visualization. 

Some gestalt principles are stronger than others. For example, the enclosure principle is much stronger than similarity. Our minds will group two objects are the enclosed before trying to group objects that are of a similar shape or color. Gaining an understanding of the strength of each principle will help in the design of your visualization.

### Principles Importance Table

| | Principle| Description|
|:---------:|--------:|--------:|
|**Powerful**|  Enclosure | providing a boundary that defines a particular region |
|**↑**| Connection | using a segment to join a set of objects together |
|**↓**| Proximity | the closeness of objects |
|**Basic level**| Similarity | a characteristic that is used to make groups homogeneous|

Please be aware that this table doesn't indicate any importance of the principles.<br>
There is arguement state that the enclosure is the strongest Gestalt principle. <br>
However, personally, I don't quite agree that each individual Gestalt principle can be ordered/organized by strength. <br>
The strength of Gestalt principles can vary depending on the context and the specific visual factor involved. Both principles play important roles in how we perceive and interpret visual information, but their influence can be influenced by factors such as the complexity of the stimuli, individual differences in perception, and the presence of other Gestalt principles at play. The rule of thumb is always to consider how you want to use the principle to support your visualization and communicate with your audience.

### Comparison: Single-Hue (Monochromatic) vs Multi-Hue Color Encoding

| Aspect | Single-Hue (Gradient / Intensity Variation) | Multiple-Hue (Distinct Colors) |
|--------|----------------------------------------------|----------------------------------|
| Best for | Ordinal data (e.g., severity, ranking, levels) | Nominal data (unordered categories) |
| Cognitive Load | Lower (less visual competition) | Higher (more color variation to process) |
| Gestalt Similarity | Shows similarity within an ordered structure | Strong categorical grouping across space |
| Perceptual Clarity | Emphasizes magnitude and order | Emphasizes category distinction |
| Scalability (Many Categories) | Scales better when categories increase | Becomes overwhelming with many categories |
| Semantic Alignment | Darker = stronger/greater (intuitive mapping) | No inherent ordering unless imposed |
| Risk | Categories may blend if contrast is weak | Visual clutter and color noise if overused |
| Legend Dependence | Often easier to interpret once order is clear | Requires stronger legend decoding |
| Use in Research Figures | Clean and elegant for structured variables | Clear when highlighting group differences |
| Example Application | Problem severity (Major → Minor → None) | Product types, political parties, regions |

### Proximity
The Gestalt principle of proximity suggests that objects that are close to one another will be interpreted as a group more than objects far away from one other.

Consider the following graph that shows the amount of air pollution in four world regions. This visualization uses proximity by putting all the India locations close to one another, all the China locations together, etc. This visualization also uses color for similarity.

Source: https://chart-studio.plotly.com/~ReadtheBox/21.embed

```python
fig, ax = plt.subplots(figsize=(12, 5))

### Title for the plot
ax.set_title('Cities with the Worst Outdoor Air Pollution PM2.5') ### your fancy title for the plot

### data specification in x and y axis

### cities on Y axis
in_cities = np.array(["Ludhiana", "Delhi", "Indore", "Bangalore"])
ch_cities = np.array(["Lanzhou", "Xining", "Urumuqi", "Beijing"])
eu_cities = np.array(["Plondiv,Bulgaria", "Krakow, Poland", "Thessaloniki, Greece", "London"])
us_cities = np.array(["Bakersfield, CA", "Riverside, CA", "Los Angeles", "New York"])

### air quality on x axis
india = np.array([251, 198, 174, 90])
china = np.array([150, 141, 140, 121])
eurp = np.array([70, 64, 56, 29])
us = np.array([38, 28, 25, 21])

b1 = ax.barh(in_cities, india, color = "green")
b2 = ax.barh(ch_cities, china, color = "red")
b3 = ax.barh(eu_cities, eurp, color = "orange")
b4 = ax.barh(us_cities, us, color = "blue")

### reverse the order
ax.invert_yaxis()  # labels read top-to-bottom

ax.legend([b1, b2, b3, b4], ["India", "China", "Eurp", "US"], loc = "right", bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(12, 6))

ax.set_title('Cities with the Worst Outdoor Air Pollution PM2.5')

# Combine city names and values
cities = (
    list(in_cities) +
    list(ch_cities) +
    list(eu_cities) +
    list(us_cities)
)

values = (
    list(india) +
    list(china) +
    list(eurp) +
    list(us)
)

# Create manual y positions with gaps
gap = 1.5
y_positions = []

current = 0
for group in [in_cities, ch_cities, eu_cities, us_cities]:
    for _ in group:
        y_positions.append(current)
        current += 1
    current += gap  # add extra space between regions

# Draw bars with region-based coloring
colors = (
    ["green"] * len(in_cities) +
    ["red"] * len(ch_cities) +
    ["orange"] * len(eu_cities) +
    ["blue"] * len(us_cities)
)

ax.barh(y_positions, values, color=colors)

ax.set_yticks(y_positions)
ax.set_yticklabels(cities)

ax.invert_yaxis()

# Custom legend
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='green', label='India'),
    Patch(facecolor='red', label='China'),
    Patch(facecolor='orange', label='Europe'),
    Patch(facecolor='blue', label='US')
]

ax.legend(handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```

```python
### this is an example of cluster algorithm
### we will cover this topic later this semester
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans

def loadDataSet():
    a = []
    f = open("../../data/B.txt",'r')
    lines = f.readlines()   #make lines a list, has index inside already
    f.close()
    for line in lines:
        b = ['', '']
        b[0],b[1] = line.split()
        a.append(b)
    dataSet = np.array(a,dtype=float)   
    return dataSet

X = loadDataSet()

# Fit KMeans clustering
KM_cluster = KMeans(n_clusters=2, random_state=42)
KM_cluster.fit_predict(X)
print(KM_cluster.labels_)
print("Figure below is the clustered result with KMeans.")

# Fit Agglomerative clustering
AG_cluster = AgglomerativeClustering(n_clusters=2, linkage='single')
AG_cluster.fit_predict(X)
print(AG_cluster.labels_)
print("Figure below is the clustered result with Agglomerative Clustering.")

# Plot the clusters side by side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("KMeans Clustering")
plt.scatter(X[:, 0], X[:, 1], c=KM_cluster.labels_, cmap='rainbow')

plt.subplot(1, 2, 2)
plt.title("Agglomerative Clustering")
plt.scatter(X[:, 0], X[:, 1], c=AG_cluster.labels_, cmap='rainbow')

plt.tight_layout()
plt.show()
```

### Enclosure

The tendency of the human mind to perceive incomplete objects as complete. When presented with a fragmented image or shape, people have a tendency to mentally fill in the missing information to perceive the object as whole. This principle illustrates how our brains seek to organize visual information into meaningful wholes, even when the information is incomplete or ambiguous. Enclosure is considered the strongest gestalt principle (i.e. our minds will use this principle first in grouping objects together).

In its simplest form, the principle of closure allows your eye to follow something like a dotted line to its end. But more complex applications are often seen in logos, like that for the World Wildlife Fund. Large chunks of the outline for the panda are missing, but your brain has no problem filling in the missing sections to see the whole animal.

### Continuation

The law of continuity posits that the human eye will follow the smoothest path when viewing lines, regardless of how the lines were actually drawn.

This continuation can be a valuable tool when the goal is to guide a visitor’s eye in a certain direction. They will follow the simplest path on the page, so make sure the most vital parts they should see fall within that path.

Since the eye naturally follows a line, placing items in a series in a line will naturally draw the eye from one item to the next. Horizontal sliders are one such example, as are related product listings on sites like Amazon.

When humans read, their eyes tend to follow shapes more than colors. The shape and structure of the text, including the arrangement of letters into words and sentences, guide the eye's movement across the page. Colors can be used to highlight or emphasize certain elements, but the overall reading process is primarily driven by the shape and structure of the text.

### Figure/Ground

The figure/ground principle is similar to the closure principle in that it takes advantage of the way the brain processes negative space.

Your brain will distinguish between the objects it considers to be in the foreground of an image (the figure, or focal point) and the background (the area on which the figures rest). Where things get interesting is when the foreground and background actually contain two distinct images, like this:
