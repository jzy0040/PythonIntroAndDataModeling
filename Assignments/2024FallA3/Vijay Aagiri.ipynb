### KMean Clustering

##### Add your KMean model in the cells below.

For this project, I chose

3 as the number of clusters. I know that the Iris dataset has three types of flowers: Setosa, Versicolor, and Virginica, so setting
𝐾
=
3 makes sense. I’m using the Euclidean distance measure, which is the default for KMeans. This way, I’m grouping flowers based on how close they are to each other in terms of their measurements.

Now, I’m setting up and training the KMeans model. I tell the model to create 3 clusters and set a random state so that I get the same results each time I run it. Then, I fit (or train) the model using the flower data (sepal and petal measurements).

```python
from sklearn import datasets
from sklearn.cluster import KMeans

# Loading the Iris dataset
iris = datasets.load_iris()
data = iris.data  # Using only the features (sepal/petal measurements) for clustering

# Setting up and train the KMeans model
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data)  # Training the model with the data
```

Once the model is trained, I ask it to predict which cluster each flower belongs to. This gives me labels (numbers) for each flower, showing which group it falls into. I also check the coordinates of each cluster’s center to understand where each group is located, and I print the first 10 predictions to get a quick look at the results.

```python
# Predicting the cluster labels for each data point
clusters = kmeans.predict(data)

# Printing out cluster centers and the first 10 predicted clusters
print("Cluster Centers:\n", kmeans.cluster_centers_)
print("First 10 Cluster Labels:\n", clusters[:10])  # Show the first 10 predictions
```

To better understand the results, I plot them. Each flower is shown as a point, with different colors for each cluster. The red ‘X’ symbols represent the cluster centers. By choosing 3 clusters, the model groups the flowers in a way that should roughly match the three species of iris flowers. However, because the model doesn’t actually know the species labels, some flowers might not end up in the expected group. I expect to see some overlap, especially between Versicolor and Virginica, since they have similar features.

```python
import matplotlib.pyplot as plt

# Plotting the clustering results
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', label='Clustered Data')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("KMeans Clustering of Iris Dataset")
plt.legend()
plt.show()
```

To see how initial positions for centroids affect the results, I rerun the KMeans model but with random starting points for the cluster centers. KMeans can give slightly different results depending on where it starts because it might find different “local” best groupings. By setting init='random', I can observe how using different starting points can sometimes lead to different cluster assignments, especially if the clusters are close together.

```python
# Running KMeans with random initial centroids
kmeans_random = KMeans(n_clusters=k, init='random', random_state=42)
kmeans_random.fit(data)
clusters_random = kmeans_random.predict(data)

# Plotting the results with random initial centroids
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=clusters_random, cmap='coolwarm', label='Clustered Data with Random Init')
plt.scatter(kmeans_random.cluster_centers_[:, 0], kmeans_random.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroids (Random Init)')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("KMeans Clustering with Different Initial Centroids")
plt.legend()
plt.show()
```

### Agglomerative Clustering

##### Add your Agglomerative model in the cells below.

For this task, I begin by selecting the "ward" linkage method for Agglomerative Clustering. Ward linkage is particularly suitable for this dataset as it minimizes the variance within each cluster, resulting in compact and well-defined groups. This method uses Euclidean distance as its default measure, which calculates the straight-line distance between points in feature space. Euclidean distance works effectively with the Iris dataset, as it allows clustering based on the similarity of features such as sepal and petal dimensions.

```python
# Importing necessary libraries
from sklearn.cluster import AgglomerativeClustering
from sklearn import datasets

# Loading the Iris dataset
iris = datasets.load_iris()
# Using only the features for clustering
data = iris.data

# Initializing the Agglomerative Clustering model with "ward" linkage
agglo = AgglomerativeClustering(n_clusters=3, linkage='ward')
# Fitting the model and predict cluster assignments
clusters = agglo.fit_predict(data)
```

After setting up the model with 3 clusters (to correspond with the three species of Iris flowers), I train the Agglomerative Clustering model and generate the cluster predictions for each data point. With Ward linkage, the model forms clusters that aim to minimize variance within each group, leading to compact clusters. I then check the first 10 cluster labels to get an initial view of how the model grouped these samples.

```python
# Displaying the first 10 cluster labels
print("First 10 Cluster Labels (Ward Linkage):\n", clusters[:10])

# Plotting the clustering results for "ward" linkage
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=clusters, cmap='viridis', label='Clustered Data')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Agglomerative Clustering of Iris Dataset (Ward Linkage)")
plt.legend()
plt.show()
```

The results from using Ward linkage generally align with the expected structure of the dataset, as the model groups the data into three clusters that correspond to the three species of iris flowers (Setosa, Versicolor, and Virginica). Ward linkage’s focus on minimizing within-cluster variance tends to produce tight, well-separated clusters. However, due to the similarity in features between some species, especially Versicolor and Virginica, a few data points might not fall exactly as expected. Overall, the results are close to expectations for an unsupervised clustering approach.

To understand the impact of different linkage methods on clustering results, I try both "complete" and "average" linkage methods. Complete linkage calculates the distance between clusters based on the farthest points in each cluster, resulting in clusters that tend to be more spread out. Average linkage, on the other hand, uses the average distance between points, producing clusters with potentially more varied shapes and sizes. Testing these different linkage methods allows me to observe how each approach uniquely groups data points.

```python
# Trying "complete" linkage
agglo_complete = AgglomerativeClustering(n_clusters=3, linkage='complete')
clusters_complete = agglo_complete.fit_predict(data)

# Plotting results for complete linkage
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=clusters_complete, cmap='coolwarm', label='Complete Linkage')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Agglomerative Clustering of Iris Dataset (Complete Linkage)")
plt.legend()
plt.show()

# Trying "average" linkage
agglo_average = AgglomerativeClustering(n_clusters=3, linkage='average')
clusters_average = agglo_average.fit_predict(data)

# Plotting results for average linkage
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=clusters_average, cmap='plasma', label='Average Linkage')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Agglomerative Clustering of Iris Dataset (Average Linkage)")
plt.legend()
plt.show()
```

I observe that the clustering results vary depending on the linkage method chosen. With "ward" linkage, the clusters are compact, as expected, due to its focus on minimizing within-cluster variance. "Complete" linkage, which uses the farthest distances between clusters, creates more spread-out groups. "Average" linkage produces clusters that vary more in size and shape, as it considers the average distances between points. These differences are not due to randomness, as Agglomerative Clustering is deterministic; instead, they stem from the unique ways each linkage method calculates distances between clusters.

In summary, Ward linkage’s compact clusters fit the Iris dataset well, while complete linkage results in more dispersed clusters, and average linkage strikes a balance. These experiments show how the choice of linkage can influence the clustering results, offering unique insights into the data structure based on each linkage method’s approach to grouping data points.
