# Gestalt Principles

Gestalt is a German word indicates shapes or forms and means “unified whole” in English. <br>
Gestalt psychology is a branch of psychology that focuses on the organization and interpretation of sensory information to represent and comprehend complex information.<br>
In general, this principle will help your viz more clear and effective.

There are varying interpretations of the exact meaning of Gestalt principles. Try to be flexible when design your visualization. 

Some gestalt principles are stronger than others. For example, the enclosure principle is much stronger than similarity. Our minds will group two objects are the enclosed before trying to group objects that are of a similar shape or color. Gaining an understanding of the strength of each principle will help in the design of your visualization.

### Principles Importance Table

| | Principle| Description|
|:---------:|--------:|--------:|
|**Strongest**|  Enclosure | providing a boundary that defines a particular region |
|**↑**| Connection | using a segment to join a set of objects together |
|**↓**| Proximity | the closeness of objects |
|**Weakest**| Similarity | a characteristic that is used to make groups homogeneous|

### Similarity
The Gestalt principle of similarity is where a characteristics such as color or shape is used to create homogeneous groupings of objects. Consider the following graphic that communicates to what degree (major/minor/not) teenagers believe each problem is among their peers.

The graph below, showing the percentage of American teenagers, ages 13 – 17, who feel certain problems are major or minor among people in their community, appeared elsewhere on NYTimes.com. The data is from a representative sample of 920 teens surveyed from September to November 2018.

Source: https://www.nytimes.com/2020/03/12/learning/whats-going-on-in-this-graph-mental-health-of-teenagers.html

```python
import numpy as np
from matplotlib import pyplot as plt


fig, ax = plt.subplots(figsize=(10, 5))

problems = ["Anxiety&Depression", "Bullying", "Drug Addition", "Drink Alcohol", "Poverty", "Teen Pregnancy", "Gangs"]
major = np.array([70, 55, 51, 45, 40, 34, 33])
minor = np.array([26, 35, 35, 39, 47, 44, 38])
none = np.array([4, 10, 13, 16, 13, 21, 29])

b1 = ax.barh(problems, major, color="red")

b2 = ax.barh(problems, minor, left = major, color="yellow")

b3 = ax.barh(problems, none, left = major + minor, color="blue")

ax.set_yticks(problems, labels = problems)
ax.invert_yaxis()  # labels read top-to-bottom

plt.legend([b1, b2, b3], ["Major", "Minor", "None"], title="Problem Importance", loc="best")

plt.show()
```

### Proximity
The Gestalt principle of proximity suggests that objects that are close to one another will be interpreted as a group more than objects far away from one other.

Consider the following graph that shows the amount of air pollution in four world regions. This visualization uses proximity by putting all the India locations close to one another, all the China locations together, etc. This visualization also uses color for similarity.

Source: https://chart-studio.plotly.com/~ReadtheBox/21.embed

```python
fig, ax = plt.subplots(figsize=(12, 5))

### Title for the plot
ax.set_title('Cities with the Worst Outdoor Air Pollution PM2.5') ### your fancy title for the plot

### data specification in x and y axis

### cities on Y axis
in_cities = np.array(["Ludhiana", "Delhi", "Indore", "Bangalore"])
ch_cities = np.array(["Lanzhou", "Xining", "Urumuqi", "Beijing"])
eu_cities = np.array(["Plondiv,Bulgaria", "Krakow, Poland", "Thessaloniki, Greece", "London"])
us_cities = np.array(["Bakersfield, CA", "Riverside, CA", "Los Angeles", "New York"])

### air quality on x axis
india = np.array([251, 198, 174, 90])
china = np.array([150, 141, 140, 121])
eurp = np.array([70, 64, 56, 29])
us = np.array([38, 28, 25, 21])

b1 = ax.barh(in_cities, india, color = "green")
b2 = ax.barh(ch_cities, china, color = "red")
b3 = ax.barh(eu_cities, eurp, color = "orange")
b4 = ax.barh(us_cities, us, color = "blue")

### reverse the order
ax.invert_yaxis()  # labels read top-to-bottom

ax.legend([b1, b2, b3, b4], ["India", "China", "Eurp", "US"], loc = "right", bbox_to_anchor=(1, 0.5))

plt.show()
```
